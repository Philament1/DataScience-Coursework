We load our data from their CSV files using `pd.read_csv`, then separate the data into a data matrix and target variable array. We also store a list of column names for use later.

```python
# Load data from CSV
training_data = pd.read_csv("nanoelectrodes_capacitance_samples.csv")
test_data = pd.read_csv("nanoelectrodes_capacitance_test.csv")

# Sort data and store as numpy arrays
X_train_pd = training_data.drop("Capacitance ($\\mu F / cm^2$)", axis=1)
X_train = X_train_pd.to_numpy()
y_train = training_data["Capacitance ($\\mu F / cm^2$)"].to_numpy()

X_test = test_data.drop("Capacitance ($\\mu F / cm^2$)", axis=1).to_numpy()
y_test = test_data["Capacitance ($\\mu F / cm^2$)"].to_numpy()

# Store list of column names
feature_names = X_train_pd.columns.tolist()
training_data
```

```python
def standardise(X, X_train=None):
    """Standardise features.

    Parameters:
        X (np.array): Feature matrix.
        X_train_ (np.array): An optional feature matrix to compute the statistics
            from before applying it to X. If None, just use X to compute the statistics.

    Returns:
        X_std (np.array): Standardised feature matrix
    """
    if X_train is None:
        X_train = X

    mu = np.mean(X_train, axis=0, keepdims=True)
    sigma = np.std(X_train, axis=0, keepdims=True)
    X_std = (X - mu) / sigma
    return X_std

X_test_std = standardise(X_test, X_train)
X_train_std = standardise(X_train)
```

We create a $R^2$ Score function and different MSE functions for use later.

```python
def r2_score(y_true, y_pred):
    """R^2 score to assess regression performance."""

    # Adjustment to avoid subtraction between (K,) and (1, K) arrays.
    y_true = y_true.reshape(y_pred.shape)
    y_bar = y_true.mean()

    ss_tot = ((y_true - y_bar)**2).sum()
    ss_res = ((y_true - y_pred)**2).sum()
    return 1 - (ss_res/ss_tot)

def mse(y_true, y_pred):
    """Sum of squared errors for use in Decision Trees."""
    return np.sum((y_true - y_pred) ** 2)

def mse_loss(y_true, y_pred):
    """Mean Squared Error."""
    y_true = y_true.reshape(y_pred.shape)
    return np.mean((y_true - y_pred) ** 2) 
```

### (1.1.1) [(index)](#index-task-111)

We train a Decision Tree regression model to predict the electrical capacity from the 12 features given in the dataset. We begin by setting hyperparameters `max_depth=10` and `min_samples_leaf=10`. Looking at the data, we see all variabes are continuous, so we do not need to consider account for categorical data. We then train the model on a single Decision Tree and evaluate the $R^2$ score and $\langle MSE\rangle$.

```python
def split_samples(X, y, col, val):
    """Split data at a value along a specific column."""
    left_mask = (X[:, col] < val)

    # Using the binary masks `left_mask`, we split X and y.
    X_l, y_l = X[left_mask, :], y[left_mask]
    X_r, y_r = X[~left_mask, :], y[~left_mask]

    return (X_l, y_l), (X_r, y_r)
```

```python
def split_value(X, y, column):
    """
    Calculate the index based on `column` with the split that minimizes the MSE.
    Arguments:
        X: training features, of shape (N, p).
        y: vector of target variables, of shape (N,).
        column: the column of the feature for calculating.
    Returns:
        (float, float): the resulted best MSE index and 
        the corresponding value used in splitting.
    """

    unique_vals = np.unique(X[:, column])

    index_val, threshold = np.inf, None

    # split the values of i-th feature and calculate the cost.
    for value in unique_vals:
        (X_l, y_l), (X_r, y_r) = split_samples(X, y, column, value)

        # if one of the two sides is empty, skip this split.
        if len(y_l) == 0 or len(y_r) == 0:
            continue
        
        # compute mse cost for this specifical value.
        new_cost = mse(y_l, np.mean(y_l)) + mse(y_r, np.mean(y_r))
        if new_cost < index_val:
            # update new best index.
            index_val, threshold = new_cost, value

    return index_val, threshold
```

```python
def split(X, y):
    """
    Choose the best feature to split according to criterion.
    Args:
        X: training features, of shape (N, p).
        y: vector of target variables, of shape (N,).
    Returns:
        (int, float): the best feature index and value used in splitting.
        If the feature index is None, then no valid split for the current Node.
    """

    # Initialize `split_column` to None, so if None returned this 
    # means there is no valid split at the current node.
    min_index_val = np.inf
    split_column = None
    split_val = np.nan

    for idx, column in enumerate(X.T):
        # skip column if samples are not seperable by that column.
        if len(np.unique(column)) < 2:
            continue
        index_val, current_val = split_value(X, y, idx)

        # if new best column is found,
        if index_val < min_index_val:

            # 1. the current minimum gini-index value,
            min_index_val = index_val
            # 2. corresponding column,
            split_column = idx
            # 3. corresponding split threshold.
            split_val = current_val

    return split_column, split_val
```

```python
def build_tree(X, y, feature_names, depth, max_depth=10, min_samples_leaf=10):

    # include a clause for the cases where
    # (i) no feature, 
    # (ii) all labels are the same,
    # (iii) depth exceed, or
    # (iv) X is too small
    if len(np.unique(y))==1 or depth>=max_depth or len(X)<=min_samples_leaf:
        return {'mean_value': np.mean(y)}

    # Find best split
    split_index, split_val = split(X, y)

    # If no valid split at this node, return prediction.
    if split_index is None:
        return {'mean_value': np.mean(y)}

    # Split samples (X, y) given column and split-value.
    (X_l, y_l), (X_r, y_r) = split_samples(X, y, split_index, split_val)
    return {
        'feature_name': feature_names[split_index],
        'feature_index': split_index,
        'value': split_val,
        'mean_value': None,
        'left': build_tree(X_l, y_l, feature_names, depth + 1, max_depth, min_samples_leaf),
        'right': build_tree(X_r, y_r, feature_names, depth + 1, max_depth, min_samples_leaf)
    }
```

```python
def train(X, y, feature_names):
    """ Train the Decision Tree on all features."""
    return build_tree(X, y, feature_names, depth=1)

# Store the trained tree for use later.
np.random.seed(seed)
trained_dt = train(X_train, y_train, feature_names)
```

```python
def find_predict(tree, x):
    """
    Predict a single sample with the fitted decision tree.
    Args:
        x: (np.array) a single sample features, of shape (p,).
    Returns:
        (int): predicted value.
    """
    if tree['mean_value'] is not None:
        return tree['mean_value']

    elif x[tree['feature_index']] < tree['value']:
        # go to left branch.
        return find_predict(tree['left'], x)
    else:
        # go to right branch.
        return find_predict(tree['right'], x)
```

```python
def predict(tree, X):
    """
    Predict regression results for X.
    Args:
        X: (np.array) testing sample features, of shape (N, p).
    Returns:
        (np.array): predicted testing sample labels, of shape (N,).
    """
    if len(X.shape) == 1:
        return find_predict(tree, X)
    else:
        return np.array([find_predict(tree, x) for x in X])
```

```python
def tree_score(tree, X_test, y_test):
    """Wrapper function to return scores for single Decision Tree."""
    y_pred = predict(tree, X_test)
    return (mse_loss(y_test, y_pred),
            r2_score(y_test, y_pred))

dt_mse_train, dt_r2_train = tree_score(trained_dt, X_train, y_train)
dt_mse_test, dt_r2_test = tree_score(trained_dt, X_test, y_test)

col_labels = [r'$\langle MSE\rangle$', r'$R^2$']
table = pd.DataFrame([[dt_mse_train, dt_r2_train],
                      [dt_mse_test, dt_r2_test]],
                     columns=col_labels,
                     index=["Train", "Test"])
table.style
```

Calculating the $\langle MSE\rangle$ and $R^2$ score of `0.75`, we see that our decision tree can capture high amount of variability of our training data. Comparing this to the results on our out of sample data (the test set), we see the out of sample $\langle MSE\rangle$ is over twice as large as our training $\langle MSE\rangle$ showing some overfitting.

### (1.1.2) [(index)](#index-task-112)

We now implement Random Forests by performing bagging on a samples. We also implement a Random Forest of $B$ trees with feature bagging, choosing to pick $1/3$ of the total number of features as per the standard rule-of-thumb. To do this, we create an additional `n_features` argument in our `split_rf` function that picks out the desired number of columns to sample from (without replacement) with feature bagging. We can use this same function for bagging without feature bagging by setting `n_features` to the total number of features.

We then use 5-fold cross-validation using the $\langle MSE\rangle$ as the performance metric to find the optimal value of $B$, the number of trees. This time our number of samples (558) is not divisible by 5, so we split into 5 unequal folds of sizes (112,112,112,111,111) to remain as equal as possible. This does mean the average $\langle MSE\rangle$ from the cross-validation is now an unequally weighted average of $\langle MSE\rangle$ however the folds are large enough in size for this to be negligible. We also shuffle the sample indices before taking folds to form a random partition of our training set.

```python
def split_rf(n_features, X, y):
    """
    Choose the best feature to split according to criterion.
    Args:
        n_features: number of sampled features.
        X: training features, of shape (N, p).
        y: vector of target variables, of shape (N,).
    Returns:
        The minimized MSE index, the best 
        feature index and value used in splitting.
    """

    # The added sampling step.
    columns = np.random.choice(list(range(X.shape[1])), n_features, replace=False)
    min_index, split_column, split_val = np.inf, 0, np.nan

    # Only scan through the sampled columns in `columns_dict`.
    for column in columns:
        # skip column if samples are not seperable by that column.
        if len(np.unique(X[:, column])) < 2:
            continue

        # search for the best splitting value for the given column.
        index_val, val = split_value(X, y, column)
        if index_val < min_index:
            min_index, split_column, split_val = index_val, column, val

    return min_index, split_column, split_val
```

```python
def build_tree_rf(n_features, X, y, feature_names, depth, max_depth=10, min_samples_leaf=10):
    """Build the decision tree according to the data.
    Args:
        X: training features, of shape (N, p).
        y: vector of target variables, of shape (N,).
        feature_names: record the name of features in X in the original dataset.
        depth: current depth for this node.
    Returns:
        a dict denoting the decision tree.
    """
    # include a clause for the cases where (i) all labels are the same,
    # (ii) depth exceed (iii) X is too small
    if len(np.unique(y)) == 1 or depth>=max_depth or len(X)<=min_samples_leaf:
        return {'mean_value': np.mean(y)}

    else:
        min_index, split_index, split_val = split_rf(n_features, X, y)

        # If index is infinity, it means that samples are not seperable
        # by the sampled features.
        if min_index == np.inf:
            return {'mean_value': np.mean(y)}

        (X_l, y_l), (X_r, y_r) = split_samples(X, y, split_index, split_val)
        return {
            'feature_name': feature_names[split_index],
            'feature_index': split_index,
            'value': split_val,
            'mean_value': None,
            'left': build_tree_rf(n_features, X_l, y_l, feature_names,
                                  depth + 1, max_depth, min_samples_leaf),
            'right': build_tree_rf(n_features, X_r, y_r, feature_names,
                                   depth + 1, max_depth, min_samples_leaf)
        }
```

```python
def train_rf(B, n_features, X, y, feature_names, max_depth=10, min_samples_leaf=10):
    """
    Build the Random Forest according to the training data.
    Args:
        B: number of decision trees.
        X: training features, of shape (N, p).
        y: vector of target variables, of shape (N,).
    """

    N = X.shape[0]
    training_indices = np.arange(N)
    trees = []
    # Store out of bag samples.
    oobs = []

    for _ in range(B):
        # Sample the training_indices (with replacement).
        sample = np.random.choice(training_indices, N, replace=True)
        # Calculate out of bag samples for use later.
        oob = list(set(training_indices) - set(sample))
        X_sample = X[sample, :]
        y_sample = y[sample]
        tree = build_tree_rf(n_features, X_sample, y_sample,
                             feature_names, depth=1,
                             max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        trees.append(tree)
        oobs.append(oob)

    return trees, oobs
```

```python
def predict_rf(rf, X):
    """
    Predict classification results for X.
    Args:
        rf: a trained random forest through train_rf function.
        X: testing sample features, of shape (N, p).
    Returns:
        predicted testing sample labels, of shape (N,).
    """

    if len(X.shape) == 1:
        # if we have one sample
        return np.mean([find_predict(tree, X) for tree in rf])
    else:
        # if we have multiple samples
        return np.array([np.mean([find_predict(tree, x) for tree in rf]) for x in X])
```

```python
def rf_score(rf, X_test, y_test):
    """Wrapper function to return scores of Random Forests."""
    y_pred = predict_rf(rf, X_test)
    return (mse_loss(y_test, y_pred),
            r2_score(y_test, y_pred))
```

```python
def cross_validation_score_rf(X, y, B, folds, n_features, max_depth=10, min_samples_leaf=10):
    """
    Perform cross-validation on our samples for a particular B.
    Args:
        X: training samples, of shape (N, p).
        y: vector of target variables, of shape (N,).
        B: number of trees to perform on.
        folds: list of folds with their indices.
        n_features: number of features to perform feature bagging on.
    Returns:
        mean MSE loss of the folds.
    """
    scores = []
    for fold in folds:
        # Calculate samples to train on.
        train_indexes = list(set(range(y_train.shape[0])) - set(fold))

        X_train_i = X_train[train_indexes, :]
        y_train_i = y_train[train_indexes]

        X_val_i = X_train[fold, :]
        y_val_i = y_train[fold]

        trees, _ = train_rf(B, n_features, X_train_i, y_train_i, feature_names,
                            max_depth=max_depth, min_samples_leaf=min_samples_leaf)

        score = mse_loss(predict_rf(trees, X_val_i), y_val_i)
        scores.append(score)
    # Return mean of MSE scores
    return np.mean(scores)
```

We run the 5-fold cross-validation below. we calculate the permuted fold indices once and use the same indices for each B we are testing on. We also use the rule-of-thumb of 1/3 of total features to perform feature bagging on. Note we use `array_split` instead of `split` as our number of samples is not divisible by 5.


```python
np.random.seed(seed)
folds_indexes = np.array_split(np.random.permutation(len(y_train)), 5)
n_features = X_train.shape[1] // 3

def find_optimal_B(X_train, y_train, X_test, y_test, B_list):
    """Find the optimal B using 5-fold cross-validation."""
    # Initialise values
    optimal_b = np.nan
    best_mse = np.inf
    mse_list = []
    
    # loop through all Bs in our list
    for B in tqdm(B_list):

        mean_score = cross_validation_score_rf(X_train, y_train, B,
                                               folds_indexes, n_features)
        if mean_score < best_mse:
            best_mse = mean_score
            optimal_b = B
        mse_list.append(mean_score)

    return optimal_b, best_mse, mse_list

fig112, ax112_CV = plt.subplots(figsize=(9,6))
ax112_CV.set(title="MSE Loss over choice of B",
             xlabel="B",
             ylabel=r"$\langle MSE \rangle$")
# We first run a loose search in steps of 5
B_list = np.arange(0, 60, 5)
B_list[0] = 1
optimal_b, best_mse, mse_list = find_optimal_B(X_train, y_train,
                                        X_test, y_test, B_list)
ax112_CV.plot(B_list, mse_list)
plt.show()

print(f"Optimal B: {optimal_b}")
```

From the above we see that $\langle MSE \rangle$ converges and plateaus quickly as $B$ increases. So to avoid overfitting and decrease runtime of code, we will take optimal $B$ to be 25.

```python
n_features = X_train.shape[1] // 3
optimal_b = 25
np.random.seed(seed)
optimal_b_rf, _ = train_rf(optimal_b, n_features,
                           X_train, y_train, feature_names)
rf_mse_train, rf_r2_train = rf_score(optimal_b_rf, X_train, y_train)
rf_mse_test, rf_r2_test = rf_score(optimal_b_rf, X_test, y_test)

np.random.seed(seed)
optimal_b_bdt, _ = train_rf(optimal_b, X_train.shape[1],
                            X_train, y_train, feature_names)

bdt_mse_train, bdt_r2_train = rf_score(optimal_b_bdt, X_train, y_train)
bdt_mse_test, bdt_r2_test = rf_score(optimal_b_bdt, X_test, y_test)

labels = ["Decision Tree", "Random\nForest", "Bagging"]

fig, (ax1_mse, ax1_r2) = plt.subplots(1, 2, figsize=(12, 5),
                                      layout='constrained')
width = 0.2
pos = np.array([0, 0.5, 1])

rects = ax1_mse.bar(pos,
                    [dt_mse_train, rf_mse_train, bdt_mse_train],
                    width,
                    label='Train')
ax1_mse.bar_label(rects, fmt='{:,.0f}', padding=3)
rects = ax1_mse.bar(pos + width,
                    [dt_mse_test, rf_mse_test, bdt_mse_test],
                    width,
                    label='Test')
ax1_mse.bar_label(rects, fmt='{:,.0f}', padding=3)

rects = ax1_r2.bar(pos,
                   [dt_r2_train, rf_r2_train, bdt_r2_train],
                   width,
                   label='Train')
ax1_r2.bar_label(rects, fmt='%.2f', padding=3)
rects = ax1_r2.bar(pos + width,
                   [dt_r2_test, rf_r2_test, bdt_r2_test],
                   width,
                   label='Test')
ax1_r2.bar_label(rects, fmt='%.2f', padding=3)

ax1_mse.set_ylabel(r'$\langle MSE \rangle$')
ax1_mse.set_xticks(pos + width/2, labels)
ax1_mse.legend(loc='upper left', ncols=2)
ax1_mse.set_ylim(0, 4200)

ax1_r2.set_ylabel(r'$R^2$ Score')
ax1_r2.set_xticks(pos + width/2, labels)
ax1_r2.legend(loc='upper left', ncols=2)
ax1_r2.set_ylim(0, 1)

ax1_mse.set_title(r"$\langle MSE\rangle$ Comparisons")
ax1_r2.set_title(r"$R^2$ Score Comparisons")

plt.show()

table = pd.DataFrame([[dt_mse_train, rf_mse_train, bdt_mse_train],
                      [dt_mse_test, rf_mse_test, bdt_mse_test],
                      [dt_r2_train, rf_r2_train, bdt_r2_train],
                      [dt_r2_test, rf_r2_test, bdt_r2_test]],
                      columns=["Decision Tree", 
                               "RF with Feature Bagging",
                               "Bagging"],
                      index=[r"Train $\langle MSE\rangle$",
                             r"Test $\langle MSE\rangle$",
                             r"Train $R^2$ Score",
                             r"Test $R^2$ Score"])
table.style
```

### (1.1.3) [(index)](#index-task-113)

We now use 5-fold cross-validation to find optimal values of `max_depth` and `min_samples_leaf`. Since we are optimising over two values, we run the cross-validation across a 2D grid of values. In the interest of computation time, we run a few preliminary tests on a smaller set of values to gauge where to run our full cross-validation on. Consequently, we will run the cross validation on a 15x15 grid of `max_depth` and `min_samples_leaf` values for `max_depth`$\in\{9,\cdots,23\}$ and `min_samples_leaf`$\in\{1,\cdots,15\}$. 

```python
def find_optimal_vals(X_train, y_train, X_test, y_test, B, depths, leaves):
    """Function to find optimal values of max_depth and min_samples_leaf"""
    # We create folds as before.
    np.random.seed(seed)
    folds_indexes = np.array_split(np.random.permutation(len(y_train)), 5)

    mse_list = np.zeros((len(depths), len(leaves)))
    optimal_vals = np.nan
    best_mse = np.inf

    # We iterate over the grid of values we are optimising on
    for x, max_depth in enumerate(tqdm(depths)):
        for y, min_samples_leaf in enumerate(leaves):

            mean_score = cross_validation_score_rf(X_train, y_train, B,
                                                   folds_indexes, n_features,
                                                   max_depth=max_depth,
                                                   min_samples_leaf=min_samples_leaf)
        
            if mean_score < best_mse:
                best_mse = mean_score
                optimal_vals = (max_depth, min_samples_leaf)
            
            mse_list[x, y] = (mean_score)

    return optimal_vals, best_mse, mse_list
```

```python
# We set up the ranges of values we will iterate over
m, n = 5, 5
o, l = 10, 1
max_depths, min_samples_leaves = np.arange(o, o+m*2, 2), np.arange(l, l+n*2, 2)

optimal_vals, best_mse_2, mse_list = find_optimal_vals(X_train, y_train,
                                                       X_test, y_test,
                                                       optimal_b, max_depths,
                                                       min_samples_leaves)
optimal_depth, optimal_leaf = optimal_vals
print(f"Optimal Depth: {optimal_depth}, Optimal Min Samples Leaf: {optimal_leaf}")

# Code to plot results
fig, ax = plt.subplots(figsize=(6,6))
mapping = ax.imshow(mse_list.T, extent=[o, o+m*2, l, l+n*2], origin='lower')
cax = fig.add_axes([ax.get_position().x1+0.01, ax.get_position().y0,
                    0.02, ax.get_position().height])
fig.colorbar(mapping, cax=cax, label=r"$\langle MSE \rangle$")
ax.set(title=r"$\langle MSE \rangle$ for Different Values of max_depth"+f"\nand min_samples_leaf",
        ylabel="min_samples_leaf",
        xlabel="max_depth")

plt.show()
```

We now estimate the importance factors of each factor using Out-Of-Bag samples from bagging using our optimal values we have just found, using MSE as the performance metric. We use the formula shown in the in-person slides given by
\begin{equation*}
    \text{Importance of feature k} = \text{MSE of OOB with feature k samples permuted} - \text{MSE of OOB}
\end{equation*}
We then express these as a percentage of the most important feature and plot them in a bar chart to compare.

```python
def importance_factors(X_train_pd, y_train, trees, oobs):

    importances = np.zeros((X_train_pd.shape[1], len(trees)))
    for imp_col, (tree, oob) in enumerate(zip(trees, oobs)):
        X_oob = X_train_pd.to_numpy()[oob]
        y_oob = y_train[oob]
        default_mse = mse_loss(predict_rf([tree], X_oob), y_oob)

        for idx, col in enumerate(X_train_pd):     
            perm_X_oob = X_oob.copy()
            perm_X_oob[:, idx] = np.random.permutation(X_oob[:, idx])
            permuted_mse = mse_loss(predict_rf([tree], perm_X_oob), y_oob)
            importances[idx, imp_col] = permuted_mse - default_mse
    
    return np.sort(np.mean(importances, axis=1))[::-1]
```

### (1.2.1) [(index)](#index-task-121)

We implement a Multi-Layer Perceptron model with 2 hidden layers (both with 50 neurons each) into an output layer with 1 neuron. Our input layer has 12 neurons since we have 12 features in our data matrix. Thus we require the following parameters,


Input layer $\rightarrow$ First hidden layer:
- Weights $\mathbf{W}^{(0)}\in\mathbb{R}^{50\times12}$
- Bias $\mathbf{b}^{(0)}\in\mathbb{R}^{50}$

First hidden layer $\rightarrow$ Second hidden layer:
- Weights $\mathbf{W}^{(0)}\in\mathbb{R}^{50\times50}$
- Bias $\mathbf{b}^{(0)}\in\mathbb{R}^{50}$

Second hidden layer $\rightarrow$ Output layer:
- Weights $\mathbf{W}^{(0)}\in\mathbb{R}^{1\times50}$
- Bias $\mathbf{b}^{(0)}\in\mathbb{R}$

Between each layer we also apply the same activation function,
\begin{equation*}
    \sigma(x)=\tan^{-1}(x)\ln(|x|+1)
\end{equation*}

```python
def dense(X, W, b):
    """Full-connected MLP layer.

    Parameters:
        X: K x h_in array of inputs, where K is the batch size and h_in if the input features dimension.
        W: h_out x h_in array for kernel matrix parametersm, where h_out is the output dimension.
        b: Length h_out 1-D array for bias parameters

    Returns:
        a: K x h_out array of pre-activations
    """
    a = (W @ X.T + b[:, None]).T
    return a
```

We can differentiate the activation function with respect to $x$ to obtain,
\begin{equation*}
    \nabla\sigma(x)=\frac{\ln(|x|+1)}{1+x^2}+\frac{\mathrm{sign}(x)\tan^{-1}(x)}{|x|+1}
\end{equation*}
Which we can indeed confirm to be smooth at $x=0$.

```python
# Activation function

def activation(a):
    """Compute activation with given function."""
    return np.arctan(a) * np.log(np.abs(a) + 1)

def grad_activation(a):
    """Compute grad of activation function."""
    grad = (np.log(np.abs(a) + 1) / (1 + a ** 2)
            + np.sign(a) * np.arctan(a) / (np.abs(a) + 1))
    return grad
```

```python
# grad MSE loss
def grad_mse_loss(y_true, y_pred):
    """Compute grad of MSE loss."""
    y_true = y_true.reshape(y_pred.shape)
    return 2 * (y_pred - y_true) / y_true.size
```

```python
class MLP:
    """Class to setup MLP with ease."""
    def __init__(self, seed=seed):
        """Initiate our MLP as defined in the question."""
        self.rng = np.random.default_rng(seed)
        self.layers = []
        self.add_layer(12, 50)
        self.add_layer(50, 50)
        self.add_layer(50, 1)

    def add_layer(self, in_dim, out_dim, activation="identity"):
        """Add fully connected layer to MLP.

        Parameters:
            in_dim (int): The output dimension of the layer.
            out_dim (int): The input dimension of the layer.
            activation (str): The activation function name.
        """
        # store each layer as a dictionary in the list, as shown in the
        # attached diagram.
        self.layers.append({
            # apply Glorot initialisation for weights.
            "W": self.rng.normal(size=(out_dim, in_dim)) * np.sqrt(2. / (in_dim + out_dim)),
            # initialise bias vector with zeros.
            "b": np.zeros(out_dim)
            # store the activation function (as string)
        })

    def predict(self, X):    
        """Compute a forward pass of our network."""
        # store pre- and post-activations in list
        forward_pass = [{"index": 0, "a": X, "h": X}]

        # iterate through hidden layers
        for k in range(1, len(self.layers)):
            # compute pre-activations
            a = dense(forward_pass[k - 1]["h"], self.layers[k - 1]["W"], self.layers[k - 1]["b"])
            forward_pass.append({"index": k, "a" : a, "h" : activation(a)})

        y_hat = dense(forward_pass[-1]["h"], self.layers[-1]["W"], self.layers[-1]["b"])
        # predicted target is output of last layer
        return y_hat, forward_pass
    
    def backpropagate(self, forward_pass, delta_output):
        """
        Backpropogate through our network returning gradients for all
        weights and biases.
        """
        delta = delta_output

        gradients = []
        #Â Create a list that will contain the gradients of all the layers.
        delta = delta_output

        # Iterate on layers backwardly, from output to input.
        # Calculate gradients w.r.t. weights and biases of each level and store in list of dictionaries.
        gradients = []
        for layer, forward_computes in reversed(list(zip(self.layers, forward_pass))):

            h = forward_computes["h"]
            a = forward_computes["a"]

            gradients.append({"W" : delta.T @ h,
                              "b" : delta.sum(axis=0)})

            # Calculate the delta for the backward layer.
            delta = grad_activation(a) * (layer["W"].T @ delta.T).T

        # Return now ordered list matching the layers.
        return list(reversed(gradients))
    
```

```python
def sgd_step(mlp, X, y, learning_rate = 1e-3):
    """Perform one step of Stochastic Gradient Descent."""
    # Compute the forward pass.
    y_hat, forward_pass = mlp.predict(X)

    # Compute the partial derivative of the loss w.r.t. to predictions `y_hat`.
    delta_output = grad_mse_loss(y, y_hat)

    # Apply backpropagation algorithm to compute the gradients of the MLP parameters.
    gradients = mlp.backpropagate(forward_pass, delta_output)

    for ind, layer in enumerate(mlp.layers):
        mlp.layers[ind]["W"] -= learning_rate * gradients[ind]["W"]
        mlp.layers[ind]["b"] -= learning_rate * gradients[ind]["b"]
```

```python
def sgd(X_train, y_train, X_test, y_test, mlp,
        learning_rate=5e-5,
        n_epochs=10,
        minibatchsize=8,
        seed=seed):
    """Optimise our MLP using Stochastic Gradient Descent."""
    # get random number generator
    rng = np.random.default_rng(seed)

    # compute number of iterations per epoch
    n_iterations = len(y_train) // minibatchsize

    # store losses
    losses_train = []
    losses_test = []

    epochs_bar = tqdm(range(n_epochs))
    for i in epochs_bar:

        # shuffle data
        p = rng.permutation(len(y_train))
        X_train_shuffled = X_train[p]
        y_train_shuffled = y_train[p]

        for j in range(n_iterations):
            # get batch
            X_batch = X_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]
            y_batch = y_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]

            # apply sgd step
            sgd_step(mlp, X_batch, y_batch,
                     learning_rate=learning_rate)

        # compute loss at the end of each epoch
        y_hat_train, _ = mlp.predict(X_train)
        losses_train.append(mse_loss(y_train, y_hat_train).squeeze())
        y_hat_test, _ = mlp.predict(X_test)
        losses_test.append(mse_loss(y_test, y_hat_test).squeeze())
        epochs_bar.set_description(f'train_loss: {losses_train[-1]:.2f}, '
                                   f'test_loss: {losses_test[-1]:.2f}, '
                                   f'train_R^2: {r2_score(y_train, y_hat_train):.2f} '
                                   f'test_R^2: {r2_score(y_test, y_hat_test):.2f} ')
    return mlp, losses_train, losses_test
```

We now train our MLP using stochastic gradient descent as our optimisation method and MSE as our loss function. We run the training for 300 epochs with a learning rate of 5e-5 and a minibatch size of 8.

```python
n_epochs = 300
mlp = MLP(seed=seed)

# train our MLP using SGD.
mlp, losses_train, losses_test = sgd(X_train_std, y_train, X_test_std, y_test,
                                     mlp, learning_rate=5e-5,
                                     n_epochs=n_epochs,
                                     minibatchsize=8,
                                     seed=seed)

# plot training progress
fig, ax = plt.subplots(figsize=(9, 6))
ax.semilogy(np.arange(1,n_epochs+1), losses_train, label="Train")
ax.semilogy(np.arange(1,n_epochs+1), losses_test, label="Test")
ax.set(title="MSE Losses vs Epochs",
       xlabel = "Epoch", ylabel = r"$\langle MSE \rangle$")
ax.legend()
plt.show()
```

### (1.2.2) [(index)](#index-task-122)

We now implement SGD with momentum as our optimisation method for our MLP given by the set of equations,
\begin{align*}
    \mathbf{g}_t:&=\nabla_{\theta}L(\theta_t;\mathcal{D}_m),\\
    \mathbf{v}_{t+1}&=\beta\mathbf{v}_t+\eta\mathbf{g}_t, \\
    \theta_{t+1}&=\theta_t-\mathbf{v}_{t+1}.
\end{align*}
Where $\theta$ are our parameters, $\beta\ge0$ is the momentum term and $\eta>0$ is the learning rate. the vector $\mathbf{v}$ stores the gradients of the previous descent step.


```python
def momentum_step(mlp, X, y, vt,
                  beta = 0, learning_rate = 5e-5):
    """Implement an SGD step with momentum."""
    # Compute the forward pass.
    y_hat, forward_pass = mlp.predict(X)

    # Compute the partial derivative of the loss w.r.t. to predictions `y_hat`.
    delta_output = grad_mse_loss(y, y_hat)

    # Apply backpropagation algorithm to compute the gradients of the MLP parameters.
    g_t = mlp.backpropagate(forward_pass, delta_output)

    for ind in range(len(mlp.layers)):
        vt[ind]["W"] = beta * vt[ind]["W"] + learning_rate * g_t[ind]["W"]
        vt[ind]["b"] = beta * vt[ind]["b"] + learning_rate * g_t[ind]["b"]
        mlp.layers[ind]["W"] -= vt[ind]["W"]
        mlp.layers[ind]["b"] -= vt[ind]["b"]

    return vt
```

```python
def momentum_sgd(X_train, y_train, X_test, y_test, mlp,
                 beta=0,
                 learning_rate=1e-3,
                 n_epochs=10,
                 minibatchsize=8,
                 seed=seed):
    """Optimise our MLP with SGD with momentum."""

    # get random number generator
    rng = np.random.default_rng(seed)

    # compute number of iterations per epoch
    n_iterations = len(y_train) // minibatchsize

    # store losses
    losses_train = []
    losses_test = []

    epochs_bar = tqdm(range(n_epochs))
    for i in epochs_bar:

        vt = []
        for layer in mlp.layers:
            vt.append({
                "W" : np.zeros_like(layer["W"]),
                "b" : np.zeros_like(layer["b"])
            })

        # shuffle data
        p = rng.permutation(len(y_train))
        X_train_shuffled = X_train[p]
        y_train_shuffled = y_train[p]

        for j in range(n_iterations):
            # get batch
            X_batch = X_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]
            y_batch = y_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]

            # apply sgd step
            vt = momentum_step(mlp, X_batch, y_batch, vt,
                               beta = beta,
                               learning_rate=learning_rate)

        # compute loss at the end of each epoch
        y_hat_train, _ = mlp.predict(X_train)
        losses_train.append(mse_loss(y_train, y_hat_train).squeeze())
        y_hat_test, _ = mlp.predict(X_test)
        losses_test.append(mse_loss(y_test, y_hat_test).squeeze())
        epochs_bar.set_description(f'train_loss: {losses_train[-1]:.2f}, '
                                   f'test_loss: {losses_test[-1]:.2f}, '
                                   f'train_R^2: {r2_score(y_train, y_hat_train):.2f} '
                                   f'test_R^2: {r2_score(y_test, y_hat_test):.2f} ')
    return mlp, losses_train, losses_test
```

```python
n_epochs = 300
momentum_mlp = MLP()

# train our MLP using SGD with momentum
momentum_mlp, losses_train_m, losses_test_m = momentum_sgd(X_train_std, y_train,
                                                           X_test_std, y_test,
                                                           momentum_mlp,
                                                           beta = 0.4,
                                                           learning_rate = 5e-5,
                                                           n_epochs=n_epochs,
                                                           minibatchsize=8)

# plot training progress
fig, ax = plt.subplots(figsize=(9, 6))
ax.semilogy(np.arange(1,n_epochs+1),losses_train, label="Train")
ax.semilogy(np.arange(1,n_epochs+1),losses_test, label="Test")
ax.semilogy(np.arange(1,n_epochs+1),losses_train_m, label="Train with momentum")
ax.semilogy(np.arange(1,n_epochs+1),losses_test_m, label="Test with momentum")
ax.set(title="MSE Losses vs Epochs for SGD with Momentum",
       xlabel = "Epoch", ylabel = r"$\langle MSE \rangle$")
ax.legend()
plt.show()
```

Comparing the SGD with momentum to regular SGD, we see that SGD with momentum converges faster in training than regular SGD and to a lower value. We can explain this using our $\mathbf{v}$ term in optimisation- since this keeps track of previous gradients, the term increases the reluctance for the descent direction to change. This counteracts sudden changes in descent direction that may be caused by any inherent randomness, resulting in a more direct optimal path (hence converging faster and to a lower value).

We now compare these results to that of linear regression, Ridge linear regression and Ridge linear regression with quadratic basis functions. We begin by extending our original dataset to include all quadratic terms,
\begin{equation*}
    \mathbf{X}'=[X_1,X_2,...,X_p,X_1^2,X_1X_2,...,X_1X_p,X_2^2,X_2X_3,...,X_2X_p,...,X_p^2]
\end{equation*}
We then perform Ridge linear regression with quadratic basis functions on penalties $\lambda\in\{0.0001,1,1000\}$ to find the parameters $\boldsymbol{\beta}\in\mathbb{R}^{91}$, where $\boldsymbol{\beta}$ includes a term for each of the 90 features in our extended feature set as well as an intercept term. To plot and compare the distribution of inferred coefficients, we standardise our extended set of features to reduce the effect of unit scaling.

We then use 5-fold cross-validation to find the best penalty term for this extended feature regression. Again, we permute sample indices before taking folds to obtain random samples. Finally then perform linear regression and Ridge linear regression on our original dataset with 12 features, using 5-fold cross-validation to find the best penalty term for Ridge regression and compare the results.

```python
# Compute the extended set of features wit quadratic basis functions
X_ext = X_train.copy()

for i, x in enumerate(X_train.T):
    for y in X_train.T[i:]:
        X_ext = np.hstack([X_ext, (x * y)[:, None]])
```

```python
def ridge_estimate(X, y, penalty):
    """
    Perform Ridge linear regression with a given penalty
    Parameters:
        X: N x D matrix of training inputs
        y: N x 1 vector of training targets/observations
    Returns:
        maximum likelihood parameters (D x 1)
    """
    # Include a column of 1s for the intercept term
    N, D = X.shape
    X_aug = np.hstack([np.ones((N,1)), X])
    N_aug, D_aug = X_aug.shape
    I = np.identity(D_aug)
    # Intercept term does not have a penalty bias
    I[0, 0] = 0
    beta_ridge = np.linalg.solve(X_aug.T @ X_aug + penalty * I, X_aug.T @ y)
    return beta_ridge
```

We plot the distributions of maximum likelihood parameters above for different regularisation terms shown. We observe that the magnitude of the components of $\boldsymbol{\beta}$ are much greater for smaller penalties $\lambda$. This makes sense as for Ridge linear regression, the penalty term controls the trade-off between bias and variance by penalising $\Vert\boldsymbol{\beta}\Vert$ hence reducing the size of the coefficients.

For $\lambda=0.0001$, our model is closest to regular linear regression on the extended feature set. Here, our maximum likelihood parameter is not penalised much for its size and we can see peaks for features for $i>12$ suggesting that quadratic term features also affect the prediction.

For larger $\lambda$, the size of $\beta$ is penalised much more. For $\lambda=1$, we see a roughly equal distribution across all components (excluding the intercept), suggesting most features equally affect the prediction.
```python
def score(X_test, y_test, betas, score_func=mse_loss):
  """Score function to evaluate regression given optimal parameters."""
  X_test_aug = np.hstack([np.ones((X_test.shape[0],1)), X_test])
  y_pred = X_test_aug @ betas
  return score_func(y_test, y_pred)
```

```python
def cross_validation_score(X_train, y_train, folds, penalty):
    """
    Perform cross-validation on our samples for a particular penalty.
    Args:
        X_train: training samples, of shape (N, p).
        y_train: vector of target variables, of shape (N,).
        folds: list of folds with their indices.
        penalty: Ridge penalty to cross-validate over.
    Returns:
        mean MSE loss of the folds.
    """
    scores = []
    for i in range(len(folds)):
        val_indexes = folds[i]
        train_indexes = list(set(range(y_train.shape[0])) - set(val_indexes))
        
        X_train_i = X_train[train_indexes, :]
        y_train_i = y_train[train_indexes]

        X_val_i = X_train[val_indexes, :]
        y_val_i = y_train[val_indexes]

        # We standardise within the cross-validation to avoid information leakage
        # from our new training set into our new test sets.
        X_val_i_std = standardise(X_val_i, X_train=X_train_i)
        X_train_i_std = standardise(X_train_i)

        beta_ridge = ridge_estimate(X_train_i_std, y_train_i, penalty)
        score_i = score(X_val_i_std, y_val_i,
                        beta_ridge)

        scores.append(score_i)

    # Return the average score
    return np.mean(scores)
```

```python
def choose_best_k(X_train, y_train, folds, penalties):
    """Run 5-fold CV for all penalties to find the best penalty."""
    penalty_scores = np.zeros(len(penalties))
    
    for i, penalty in enumerate(penalties):
        penalty_scores[i] = cross_validation_score(X_train, y_train, folds, penalty)

    best_pen_idx = np.argmin(penalty_scores)
    return (penalties[best_pen_idx], penalty_scores.min()), penalty_scores
```

We run 5-fold cross-validation with similar setup to before to find optimal penalty. We use `np.logspace` to create our range of  $\lambda$ for cross-validation. We decide to pick values of $\lambda$ betwen 0.0001 and 1000 as it seems to capture the best value of $\lambda$ in its range.

```python
def lin_reg_scores(X_train, X_test, y_train, y_test, penalty):
    """Wrapper function to return generalisation power scores."""
    beta = ridge_estimate(X_train, y_train, penalty)
    mse_train = score(X_train, y_train, beta)
    mse_test = score(X_test, y_test, beta)
    r2_train = score(X_train, y_train, beta,
                     score_func=r2_score)
    r2_test = score(X_test, y_test, beta,
                    score_func=r2_score)

    return (mse_train, r2_train), (mse_test, r2_test)
```

## Data Selection Choices

Looking at the data given for the question, we see a "Patient ID" column which simply indexes all our data. This does not seem like a useful data feature to use as a predictor for our tasks as it does not seem to encode any useful information about the type of brain tumor. Moreover, we have a column called "Location" which seems to contain discrete data from 0 to 4 pointing to a specific location in the brain. Since it is not obvious whether this numbering is arbitrary, we will decide to treat this column as a discrete variable and use the Hamming distance as a distance metric (In this case the output will either be 0 or 1).

We now require a combined distance function for our KNN since we now have a mix of discrete and continuous variables for our columns. We choose to simply add the distances together since upon checking, both distances seem to be of roughly the same order of magnitude so the sum is representative of both values. For a more refined method, we could create an additional hyperparameter for the weighted sum of these two distances, and run a cross validation to find the best weighting.

## Task Details

Other than this distance function, we implement kNN and Cross Validation the same as in the labs. As stated in the question, we use five-fold cross validation to find the optimal k. This time the number of samples is exactly divisble by 5, so we obtain 5 equally sized folds. Again, we choose to shuffle the sample indices before folding to choose a random split.

We then implement the accuracy and precision metrics used in the question. We define the micro-average accuracy to be the fraction of correct predictions over total predictions, which turns out to be the same as the micro-average precision. We define the macro-averages as the average of the individual metrics for each class and the weighted-average as the weighted average.

For instance, macro-average accuracy = (class 0 accuracy + class 1 accuracy + class 2 accuracy) / 3. For each class we have Accuracy = (TP + TN) / Total Predictions, whilst treating the class as a binary outcome. Since the denominator is the same, we can simplify this to macro-average accuracy = (total TP + total TN) / (3 * Total Predictions). Now total TP is the sum of the diagonals in the confusion matrix, and total TN is the sum of the diagonals + sum of all entries (Total Predictions). Since the sum of all diagonals is the same as the number of true classifications, we have the formula
\begin{equation*}
    \text{Macro-average Accuracy} = \frac{1}{3} + \frac{2}{3}\text{Micro-average Accuracy}
\end{equation*}
Macro-average precision is defined much the same way, by averaging the precision for each class by TP/(TP+FP). In the weighted averages, we use the class frequencies of the test set data. This is because we want the metric to be representative of the true dataset we are testing on.

```python
def euc_dist(u, v):
    """Return the Euclidean distance."""
    return np.sqrt(np.sum((u - v) ** 2, axis=1))

def hamming(u, v):
    """Return the Hamming distance."""
    return (u != v).astype(int)

def combined_dist(X_train, x_test):
    """Return a combined distance for kNN."""
    euc = euc_dist(np.delete(X_train, cat_col, 1),
                   np.delete(x_test, cat_col))
    ham = hamming(X_train[:, cat_col], x_test[cat_col])
    return euc + ham
```

```python
def k_neighbours(X_train, X_test, k=5):
    """Find k nearest neighbours."""
    n_neighbours = k
    neigh_ind = []

    # compute distance from each point x_text in X_test to all points in X_train
    point_dist = [combined_dist(X_train, x_test) for x_test in X_test]
    
    neigh_ind = np.argsort(point_dist, axis=1)[:,:k]

    return np.array(neigh_ind)
```

```python
def predict(X_train, y_train, X_test, k=5):
  """Predict using kNN."""
  neighbours = k_neighbours(X_train, X_test, k=k)
  y_pred =  np.array([np.argmax(np.bincount(y_train[neighbour]))
                      for neighbour in neighbours])
  return y_pred
```

We now implement functions to calculate the metrics we mentioned before. We use the function class_qualities to store accuracies and precisions per class by calculating values from the confusion matrix.

```python
def class_qualities(y_true, y_pred):
    """Compute class-based accuracy and precision."""
    classes = [0, 1, 2]
    accuracies = np.zeros(3)
    precisions = np.zeros(3)
    N = y_true.size

    for c in classes:
        true_c = y_true[y_true == c].size
        pred_c = y_pred[y_pred == c].size
        matches = y_true[y_pred == y_true]
        TP = matches[matches == c].size
        FP = pred_c - TP
        FN = true_c - TP
        TN = y_true.size - TP - FP - FN
        accuracies[c] = (TP + TN) / N
        if TP == 0 and FP == 0:
            precisions[c] = np.nan
        else:
            precisions[c] = TP / (TP + FP)

    return accuracies, precisions

def accuracy(y_true, y_pred):
    """Compute micro, macro and class-weighted average accuracy"""
    class_acc, _ = class_qualities(y_true, y_pred)
    N = y_true.size
    micro = np.sum(y_true == y_pred) / N
    weights = np.bincount(y_test) / N
    weighted = np.sum(class_acc * weights)

    # Averaging all accuracies for each class we obtain,
    macro = 1/3 + 2/3 * micro
    return micro, macro, weighted

def precision(y_true, y_pred):
    """Compute micro, macro and class-weighted average precision"""
    _, class_prc = class_qualities(y_true, y_pred)
    N = y_true.size
    micro = np.sum(y_true == y_pred) / N
    weights = np.bincount(y_true) / N
    macro = np.mean(class_prc)
    weighted = np.sum(class_prc * weights)
    return micro, macro, weighted

def score(X_train, y_train, X_test, y_test, k=5):
    """Wrapper function to compute qualities"""
    y_pred = predict(X_train, y_train, X_test, k)
    score, _, _ = accuracy(y_test, y_pred)
    return score
```

```python
def cross_validation_score(X_train, y_train, folds, k):
    """
    Perform cross-validation on our samples for a particular k in kNN.
    Args:
        X_train: training samples, of shape (N, p).
        y_train: vector of target variables, of shape (N,).
        folds: list of folds with their indices.
        k: number of neighbours to cross-validate over.
    Returns:
        mean micro-averaged accuracy of the folds.
    """
    scores = []
    for i in range(len(folds)):
        val_indexes = folds[i]
        train_indexes = list(set(range(y_train.shape[0])) - set(val_indexes))

        X_train_i = X_train[train_indexes, :]
        y_train_i = y_train[train_indexes]

        X_val_i = X_train[val_indexes, :]
        y_val_i = y_train[val_indexes]

        X_val_i = standardise(X_val_i, X_train=X_train_i)
        X_train_i = standardise(X_train_i)

        score_i = score(X_train_i, y_train_i, X_val_i, y_val_i, k=k)

        scores.append(score_i)

    return np.mean(scores)
```

```python
def choose_best_k(X_train, y_train, folds, k_range):
    """Choose best k for kNN using cross-validation"""
    k_scores = np.zeros((len(k_range),))
  
    for i, k in enumerate(k_range):
        k_scores[i] = cross_validation_score(X_train, y_train, folds, k)

    best_k_index = np.argmax(k_scores)
    return k_range[best_k_index], k_scores
```

We now run five-fold cross validation on our kNN classifier. Our values of k must be integers, so we set a range of values from 1 to 30. After ~k=20, the micro-averaged accuracy scores decrease so we do not need to consider larger k as we begin to generalise too much.

### (2.1.2) [(index)](#index-task-212)

To design a weighted version of kNN to better predict the minority class, we will add weightings to the prediction of the most occurring class out of the k neighbours. Since we want to better predict the minority class, we want to increase the weighting of the minority class in selecting the most occurring class. To do this, we divide by the class frequencies of our values in the training set as this will weight the minority classes higher and run kNN as usual otherwise.

```python
def weighted_predict(X_train, y_train, X_test, k=5):
    """Weighted implemention of kNN."""
    neighbours = k_neighbours(X_train, X_test, k=k)
    weights = np.bincount(y_train, minlength=3)
    y_pred =  np.array([np.argmax(np.bincount(y_train[neighbour], minlength=3)
                                  / weights) # divide by class count
                        for neighbour in neighbours])
    return y_pred
```

```python
def weighted_comparison(X_train, y_train, X_test, y_test, k=5):
    """Wrapper function to compute scores comparing methods."""
    y_pred = weighted_predict(X_train, y_train, X_test, k)
    micro_acc_w, macro_acc_w, _ = accuracy(y_test, y_pred)
    micro_prc_w, macro_prc_w, _ = precision(y_test, y_pred)
    y_pred_w = predict(X_train, y_train, X_test, k)
    micro_acc_n, macro_acc_n, _ = accuracy(y_test, y_pred_w)
    micro_prc_n, macro_prc_n, _ = precision(y_test, y_pred_w)

    return ({
                'normal' : (micro_acc_n, macro_acc_n),
                'weighted' : (micro_acc_w, macro_acc_w)
            },
            {
                'normal' : (micro_prc_n, macro_prc_n),
                'weighted' : (micro_prc_w, macro_prc_w)
            })
```

### (2.1.3) [(index)](#index-task-213)

We now train a 2-step kNN algorithm by first classifying whether the tumor is bengign (class 0) or malignant (classes 1 or 2), by combining classes 1 and 2 together. Then if we predict a malignant tumor, we run a second kNN with $k=1$ to predict whether th tumor is a glioma type (class 1) or meningioma type (class 2). This is all implemented in the function two_step_predict below. For efficiency, we keep track of indices where the first prediction is malignant and then replace these values with 1 or 2 in-place.

```python
def two_step_predict(X_train, y_train, X_test, k=5):
    """Implement 2-step kNN as described above."""
    # Turn into binary classification task.
    malignant = (y_train != 0)
    neighbours = k_neighbours(X_train, X_test, k=k)
    y_combined = y_train.copy()
    y_combined[y_train != 0] = 1
    y_pred =  np.array([np.argmax(np.bincount(y_combined[neighbour], minlength=3))
                        for neighbour in neighbours])
    # Locate non-zero indices
    non_zero_pred = (y_pred != 0)
    if non_zero_pred.any():
        # Perform second kNN on malignant tumors
        neighbours = k_neighbours(X_train[malignant], X_test[non_zero_pred], k=1)

        y_pred_mal = np.array([np.argmax(np.bincount(y_train[malignant][neighbour]))
                                    for neighbour in neighbours])
        # Replace malignant tumors with predicted classes.
        y_pred[non_zero_pred] = y_pred_mal
    return y_pred
```

```python
def two_step_qualities(X_train, y_train, X_test, y_test, k=5):
    """Wrapper function to compute qualities for comparison."""
    y_pred = two_step_predict(X_train, y_train, X_test, k)
    micro_acc, macro_acc, _ = accuracy(y_test, y_pred)
    micro_prc, macro_prc, _ = precision(y_test, y_pred)
    return (micro_acc, macro_acc), (micro_prc, macro_prc)
```

From our results above, we see the 2-step kNN has higher accuracy and precision metrics than the previous two methods. The reason for this may be due to the underlying structure in our classification problem- a malignant vs benign tumor may in general be clustered further apart, whilst within the malignant tumors class 1 and class 2 may be clustered closer together. In such a case, it would be beneficial to use two different values of k in our kNN as we need to consider different scales of distance. Thus, a 2-step classifier could prove much more useful than a single kNN as it introduces another hyerparameter to tune.

Moreover, we can in general view class 1 and class 2 as subclasses of malignant tumors, which may result in them sharing similarities between them that are not exhibited in class 0 (benign) tumors. Thus it would make sense to use a 2-step classification to separate out and compare these differences more directly.

## (2.2) Logistic regression vs kernel logistic regression [(index)](#index-task-22)

### (2.2.1) [(index)](#index-task-221)

We form the binary classification task of predicting a benign vs malignant tumor like we did at the start of the 2-step kNN. We now use a penalised logistic regression model given in the question to train on. The loss function is like a mean log-liklihood function from our usual logistic regression with an additional penalty term. We further note that the intercept term, $\beta_0$ is kept separate from $\boldsymbol{\beta}$ as we do not augment our data matrix with a column of ones.

To perform the logistic regression, we need to find the derivatives of the loss function with respect to $\boldsymbol{\beta}$ and $\beta_0$. We can compute the derivatives in a similar fashion to the ordinary logistic regression to obtain,

\begin{align*}
    \nabla_{\boldsymbol{\beta}}E(L)&=\frac{1}{N}\sum_{i=1}^N(h_{\boldsymbol{\beta},\beta_0}(\boldsymbol{X}^{(i)})-y^{(i)})\boldsymbol{X}^{(i)} \\
    \nabla_{\beta_0}E(L)&=\frac{1}{N}\sum_{i=1}^N(h_{\boldsymbol{\beta},\beta_0}(\boldsymbol{X}^{(i)})-y^{(i)})
\end{align*}

Here, $\boldsymbol{X}^{(i)}$ the column vector of row $i$ of matrix $\boldsymbol{X}$. We then implement logistic regression as seen before, keeping track of and updating both variables $\boldsymbol{\beta}$ and $\beta_0$.

```python
def loss(X, y, beta, beta0, lam=0.0025):
    """Compute the penalised loss function in the question."""
    y_log = h(X, beta, beta0)
    return (-np.mean(y * np.log(y_log)
                     + (1 - y) * np.log(1 - y_log))
            + lam / 2 * np.linalg.norm(beta) ** 2)

def h(X, beta, beta0):
    """Compute transformed logistic map."""
    return 1 / (1 + np.exp(-(X @ beta + beta0)))

def grad_loss(X, y, beta, beta0, lam=0.0025):
    """Compute the gradient of the loss function."""
    y_log = h(X, beta, beta0)
    beta_loss = np.mean(X.T * (y_log - y), axis=1) + lam * beta
    beta0_loss = np.mean(y_log - y)
    return beta_loss, beta0_loss
```

```python
def propagate(X, y, beta, beta0):
    """Propogate grads in each step."""
    # Mean sample loss function
    mean_loss = loss(X, y, beta, beta0)

    # Derivatives
    beta_loss, beta0_loss = grad_loss(X, y, beta, beta0)
    mean_loss = np.squeeze(mean_loss)

    return beta_loss, beta0_loss, mean_loss
```

```python
def optimise(X, y, beta, beta0, num_iterations=1000, learning_rate=0.1, print_loss=False):
    """
    Optimise parameters for logistic regression.
    Parameters:
        X: Data matrix (N, p)
        y: Target classes (N,)
        beta: Parameter to optimise (p,)
        beta0: Intercept
    Returns:
        Optimised set of parameters.
    """
    mean_loss_history = []
    
    for i in range(num_iterations):

        # Calculating the loss and gradients
        dbeta, dbeta0, mean_loss = propagate(X, y, beta, beta0)
      
        # Updating procedure
        beta -= learning_rate * dbeta
        beta0 -= learning_rate * dbeta0
      
        # Record the loss values
        if i % 10 == 0:
            mean_loss_history.append(mean_loss)
      
        # Printing the loss every 200 iterations
        if print_loss and i % 1000 == 0:
            print ('Mean loss after iteration %i: %f' %(i, mean_loss))

    # Saving parameters and gradients in dictionary
    params = {'betas': (beta, beta0)}
    grads = {'dbeta': (dbeta, dbeta0)}
  
    return params, grads, mean_loss_history
```

```python
def model(X_train, y_train, X_test, y_test,
          num_iterations=1000, learning_rate=0.1, print_loss=False):
    """
    Run logistic regression model.
    Parameters:
        X_train: Data matrix (N, p)
        y_train: Target classes (N,)
        X_test: Test set matrix (M, p)
        y_test: Target test classes (M,)
    Returns:
        Trained model with predicted values and optimised parameters.
    """
    # Initialising parameters with zeros
    beta, beta0 = np.zeros(X_train.shape[1]), 0.
    # last element is beta0

    # Gradient descent
    parameters, grads, mean_loss_history = optimise(X_train, y_train, beta, beta0,
                                                    num_iterations, learning_rate,
                                                    print_loss=print_loss)

    # Retrieving parameter vector beta from dictionary 'parameters'
    beta, beta0 = parameters['betas']

    # Predicting test and training set examples
    y_pred_test = np.round(h(X_test, beta, beta0))
    y_pred_train = np.round(h(X_train, beta, beta0))

    # Printing train/test accuracy
    print('Training accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_train - y_train)) * 100))
    print('Test accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_test - y_test)) * 100))

    # Saving all the information
    d = {'mean_loss_history': mean_loss_history, 'betas': (beta, beta0),
        'y_pred_test': y_pred_test, 'y_pred_train': y_pred_train}

    return d
```

```python
# Create binary classification vectors
y_train_binary = y_train.copy()
y_test_binary = y_test.copy()
y_train_binary[y_train != 0] = 1
y_test_binary[y_test != 0] = 1

# Store accuracies for later
train_acc = np.zeros(3)
test_acc = np.zeros(3)

# Run model
d = model(X_train_std, y_train_binary, X_test_std, y_test_binary,
          num_iterations=7000, learning_rate=0.1, print_loss=True)

train_acc[0] = 1 - np.mean(np.abs(d['y_pred_train'] - y_train_binary))
test_acc[0] = 1 - np.mean(np.abs(d['y_pred_test'] - y_test_binary))
```

We run the model with 5000 iterations to compare to the kernelised version later. After around 3000 iterations the mean loss stabilises to 3d.p. We see the final training accuracy is $77.9\%$ whilst the test accuracy is $83.3\%$. We also plot the mean loss over iterations below.

We now want to plot the precision-recall curve for our predictions, done in the pr-curve fuction below. Similar to plotting the ROC curve in the labs, we use each unique point in our set of predictions as a threshold, and calculate the precision and recall for each threshold. We also calculate the area under the curve (AUC) using np.trapz.

```python
def pr_curve(y_true, y_log):
    """
    Arguments:
    y_true: Ground truth labels with size N
    y_log: Probabilities produced by logistic regression model with size N
  
    Returns:
    auc: Area Under the Curve (AUC)
    tpr_l: List of true positive rate (TPR) values for each scanned threshold
    fpr_l: List of false positive rate (FPR) values for each scanned threshold
    thresholds: Scanned thresholds sorted in decreasing order
    
    """
    # List of distinct values in y_log, sorted sorted in decreasing order
    thresholds = reversed(sorted(set(y_log)))
    tpr_l, prc_l = [], []

    for threshold in thresholds:
        # Thresholding
        y_thresholded = (y_log >= threshold)

        # True positives
        tp = np.sum(y_true & y_thresholded)
        # False positives
        fp = np.sum((~y_true) & y_thresholded)
        # False negatives
        fn = np.sum(y_true & (~y_thresholded))

        tpr = tp / (tp + fn)
        prc = tp / (tp + fp)
        tpr_l.append(tpr)
        prc_l.append(prc)

    # Compute AUC using Trapezoidal rule
    auc = np.trapz(prc_l, tpr_l)
    return auc, tpr_l, prc_l, thresholds
```

```python
# Recover predictions
beta, beta0 = d['betas']
y_log = h(X_test_std, beta, beta0)
auc, tpr, prc, th = pr_curve(y_test_binary, y_log)
# Plot Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.title(f'Precision Recall Curve')
plt.plot(tpr, prc, 'b', label=f'AUC = {auc:0.2f}')
plt.legend(loc='upper right')
plt.ylabel('Precision')
plt.xlabel('Recall')
plt.show()
```

Above we have plotted the precision-recall curve for our normal logistic regression model. Since the curve lies entirely above the line $\texttt{Precision}=0.5$, we have that the model certainly predicts better than random. Moreover, $AUC=0.86$ is high further supporting that we have a good prediction model, as it indicates a high recall and precision.

### (2.2.2) [(index)](#index-task-222)

We set up the problem with the data matrix $\boldsymbol{X}\in\mathbb{R}^{N\times p}$ and classification vectors $\boldsymbol{y}\in\mathbb{R}^p$. We represent the kernel as $k(\boldsymbol{x},\boldsymbol{y})=\phi(\boldsymbol{x})\cdot\phi(\boldsymbol{y})$. We now optimise the kernelised logistic regression in terms of $\boldsymbol{\beta}_{\phi(\boldsymbol{X})}\in\mathbb{R}^N$, a higher dimension set of parameters on our new set of features, $\phi(\boldsymbol{X})\in\mathbb{R}^N$. This allows us to rewrite the loss function as,
\begin{equation*}
    E(L) = -\frac{1}{N}\sum_{i=1}^{N}y^{(i)}\log h_{\boldsymbol{\beta}_{\phi(\boldsymbol{X})},\beta_0}(\phi(\boldsymbol{X}^{(i)}))+(1-y^{(i)})\log(1-h_{\boldsymbol{\beta}_{\phi(\boldsymbol{X})}, \beta_0}(\phi(\boldsymbol{X}^{(i)}))) + \frac{\lambda}{2}\Vert\boldsymbol{\beta}_{\phi(\boldsymbol{X})}\Vert^2
\end{equation*}
With the function $h_{\boldsymbol{\beta},\beta_0}(\boldsymbol{x})=h(\boldsymbol{\beta},\beta_0,\boldsymbol{x})$ as defined as before. Now if we introduce a vector $\boldsymbol{u}\in\mathbb{R}^N$ such that,
\begin{equation*}
    \boldsymbol{\beta}_{\phi(\boldsymbol{X})} = \sum_{i=1}^N\phi(\boldsymbol{X}^{(i)})\boldsymbol{u}_i
\end{equation*}
We have the simplifications,
\begin{equation*}
    \phi(\boldsymbol{X}^{(i)})^T\boldsymbol{\beta}_{\phi(\boldsymbol{X})}=\boldsymbol{K}^{(i)T}\boldsymbol{u},\qquad \Vert\boldsymbol{\beta}_{\phi(\boldsymbol{X})}\Vert^2=\boldsymbol{\beta}_{\phi(\boldsymbol{X})}\cdot\boldsymbol{\beta}_{\phi(\boldsymbol{X})}=\boldsymbol{u}^T\boldsymbol{Ku}
\end{equation*}
where $\boldsymbol{K}$ is the Gram matrix with entries, $\boldsymbol{K}_{ij}=\phi(\boldsymbol{X}^{(i)})\cdot\phi(\boldsymbol{X}^{(j)})=k(\boldsymbol{X}^{(i)},\boldsymbol{X}^{(j)})$. This lets us write the optimisation problem in terms of $\boldsymbol{K}$ while optimising $\boldsymbol{u}$ and $\beta_0$,
\begin{equation*}
    E_K(L) = -\frac{1}{N}\sum_{i=1}^{N}y^{(i)}\log h_{\boldsymbol{u},\beta_0}(\boldsymbol{K}^{(i)})+(1-y^{(i)})\log(1-h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})) + \frac{\lambda}{2}\boldsymbol{u}^T\boldsymbol{Ku}
\end{equation*}
We can differentiate this with respect to $\boldsymbol{u}$ and $\beta_0$ in a similar fashion to the non-kernelised logistic regression to obtain derivatives,
\begin{align*}
    \nabla_{\boldsymbol{u}}E_K(L) &= \frac{1}{N}\sum_{i=1}^N\left((h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})-y^{(i)})\boldsymbol{K}^{(i)}\right) + \lambda\boldsymbol{Ku} \\
    \nabla_{\beta_0}E_K(L) &= \frac{1}{N}\sum_{i=1}^N(h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})-y^{(i)})
\end{align*}
To indeed show this is a convex optimisation problem, it is enough to show the Hessian of the new loss function $E_K$ is positive semi-definite. We begin by taking second partial derivatives and noting $\boldsymbol{K}$ is symmetric,
\begin{align*}
    \frac{\partial^2 E_K}{\partial\boldsymbol{u}\partial\boldsymbol{u}} &= \frac{1}{N}\sum_{i=1}^N h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})(1-h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)}))\boldsymbol{K}^{(i)}\boldsymbol{K}^{(i)T}+\lambda\boldsymbol{K} \\
    \frac{\partial^2 E_K}{\partial\boldsymbol{u}\partial\beta_0} &= \frac{1}{N}\sum_{i=1}^N h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})(1-h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)}))\boldsymbol{K}^{(i)} \\
    \frac{\partial^2 E_K}{\partial\beta_0\partial\beta_0} &= \frac{1}{N}\sum_{i=1}^N h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})(1-h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)}))
\end{align*}
Allowing us to write the Hessian in the form,
\begin{equation*}
    \nabla^2_{\boldsymbol{u},\beta_0} E_K = \frac{1}{N}\sum_{i=1}^N \begin{pmatrix}
        \gamma_i\boldsymbol{K}^{(i)}\boldsymbol{K}^{(i)T} & \gamma_i\boldsymbol{K}^{(i)} \\
        \gamma_i\boldsymbol{K}^{(i)T} & \gamma_i
    \end{pmatrix}
    +
    \begin{pmatrix}
    \boldsymbol{K} & \boldsymbol{0} \\ \boldsymbol{0} & 0
    \end{pmatrix}
\end{equation*}
where $\gamma_i=h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})(1-h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)}))$. Since the sum of positive semi-definite matrices is positive semi-definite we just need to show each term in the sum and the final matrix are positive semi-definite. Now let $(\boldsymbol{v}, w)\in\mathbb{R}^{N+1}$ be an arbitrary vector. since $h_{\boldsymbol{u}, \beta_0}(\boldsymbol{K}^{(i)})\in(0,1)$, we have each $\gamma_i\ge0$. Now,
\begin{align*}
    (\boldsymbol{v}^T,w)\begin{pmatrix}
        \gamma_i\boldsymbol{K}^{(i)}\boldsymbol{K}^{(i)T} & \gamma_i\boldsymbol{K}^{(i)} \\
        \gamma_i\boldsymbol{K}^{(i)T} & \gamma_i
    \end{pmatrix}(\boldsymbol{v},w) &= \gamma_i\boldsymbol{v}^T\boldsymbol{K}^{(i)}\boldsymbol{K}^{(i)T}\boldsymbol{v}+\gamma_iw\boldsymbol{v}^T\boldsymbol{K}^{(i)} + \gamma_iw\boldsymbol{K}^{(i)T}\boldsymbol{v}+\gamma_iw^2 \\
    &= \gamma_i(\boldsymbol{K}^{(i)T}\boldsymbol{v}+w)^T(\boldsymbol{K}^{(i)T}\boldsymbol{v}+w) \\
    &= \gamma_i\Vert\boldsymbol{K}^{(i)T}\boldsymbol{v}+w\Vert^2\ge0
\end{align*}
We can also write $\boldsymbol{K}=\boldsymbol{\phi}\boldsymbol{\phi}^T$ where $\boldsymbol{\phi}=(\phi(\boldsymbol{X^{(1)}}),\cdots,\phi(\boldsymbol{X^{(N)}})^T$. So,
\begin{align*}
    (\boldsymbol{v}^T,w)
    \begin{pmatrix}
    \boldsymbol{K} & \boldsymbol{0} \\ \boldsymbol{0} & 0
    \end{pmatrix}
    (\boldsymbol{v},w) &= \boldsymbol{v}^T\boldsymbol{Kv} = (\boldsymbol{\phi}^T\boldsymbol{v})^T\boldsymbol{\phi}^T\boldsymbol{v} \\
    &= \Vert\boldsymbol{\phi}^T\boldsymbol{v}\Vert^2 \ge 0
\end{align*}
Since all matrices involved are positive semi-definite, we must have the Hessian, $\nabla^2_{\boldsymbol{u},\beta_0} E_K$, must also be positive semi-definite. Hence this is a convex optimisation problem.

```python
def kernel_matrix(X1, X2, a):
    """
    Calculate the Kernel matrix, K_ij=k(X_i,X_j)
    a = alpha parameter in laplacian kernel
    """
    n, _ = X1.shape
    m, _ = X2.shape
    kernel = np.zeros((n, m))
    
    for ind, x1 in enumerate(X1):
        val = np.exp(-a * np.linalg.norm(X2 - x1, ord=1, axis=1))
        kernel[ind, :] = val
    return kernel

def kernel_loss(u, K, y, beta0, lam=0.0025):
    """Evaluate new loss function."""
    k_y_log = h(K, u, beta0)
    return (-np.mean(y * np.log(k_y_log)
                     + (1 - y) * np.log(1 - k_y_log))
            + lam / 2 * u.T @ K @ u)

def grad_kernel_loss(u, K, y, beta0, lam=0.0025):
    """Evaluate new derivatives of loss function."""
    # grad w.r.t u, beta0
    k_y_log = h(K, u, beta0)
    beta_loss = np.mean(K.T * (k_y_log - y), axis=1) + lam * K @ u
    beta0_loss = np.mean(k_y_log - y)

    return beta_loss, beta0_loss
```

### (2.2.3) [(index)](#index-task-223)

Similar to before we implement the kernelised logistic regression via gradient descent, using the formulations we derived in the previous part. We perform this on $\alpha=100$ and $\alpha=0.3$.

```python
def kernel_optimise(u, K, y, beta0, num_iterations=1000, learning_rate=0.1, print_loss=False):
    """
    Optimise parameters for kernelised logistic regression.
    Parameters:
        u: Parameter to optimise (N,)
        K: Kernel matrix (N, N)
        y: Target classes (N,)
        beta0: Intercept
    Returns:
        Optimised set of parameters.
    """
    mean_loss_history = []
    
    for i in range(num_iterations):

        # Calculating the loss and gradients
        du, dbeta0 = grad_kernel_loss(u, K, y, beta0)
        mean_loss = np.squeeze(kernel_loss(u, K, y, beta0))
      
        # Updating procedure
        u -= learning_rate * du
        beta0 -= learning_rate * dbeta0
      
        # Record the loss values
        if i % 10 == 0:
            mean_loss_history.append(mean_loss)
      
        # Printing the loss every 100 iterations
        if print_loss and i % 1000 == 0:
            print ('Mean loss after iteration %i: %f' %(i, mean_loss))

    # Saving parameters and gradients in dictionary
    params = (u, beta0)
    grads = (du, dbeta0)
  
    return params, grads, mean_loss_history
```

```python
def kernel_model(X_train, y_train, X_test, y_test,
                 alpha=0.3, num_iterations=2000,
                 learning_rate=0.1, print_loss=False):
     """
     Run kernelised logistic regression model.
     Parameters:
          X_train: Data matrix (N, p)
          y_train: Target classes (N,)
          X_test: Test set matrix (M, p)
          y_test: Target test classes (M,)
     Returns:
          Trained model with predicted values and optimised parameters.
     """
     # Initialising parameters with zeros
     K = kernel_matrix(X_train, X_train, alpha)
     u, beta0 = np.zeros(K.shape[1]), 0.

     # Gradient descent
     parameters, grads, mean_loss_history = kernel_optimise(u, K, y_train, beta0,
                                                            num_iterations, learning_rate,
                                                            print_loss=print_loss)

     # Retrieving parameter vector beta from dictionary 'parameters'
     u, beta0 = parameters

     # Predicting test and training set examples
     K_test = kernel_matrix(X_test, X_train, alpha)
     y_pred_test = np.round(h(K_test, u, beta0))
     y_pred_train = np.round(h(K, u, beta0))

     # Printing train/test accuracy
     print('Training accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_train - y_train)) * 100))
     print('Test accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_test - y_test)) * 100))

     # Saving all the information
     d = {'mean_loss_history': mean_loss_history, 'params': (u, beta0),
          'y_pred_test': y_pred_test, 'y_pred_train': y_pred_train}
     
     return d
```
