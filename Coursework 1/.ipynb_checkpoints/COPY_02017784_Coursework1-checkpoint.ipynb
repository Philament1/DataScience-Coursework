```python
training_data = pd.read_csv("nanoelectrodes_capacitance_samples.csv")
test_data = pd.read_csv("nanoelectrodes_capacitance_test.csv")

X_train_pd = training_data.drop("Capacitance ($\\mu F / cm^2$)", axis=1)
X_train = X_train_pd.to_numpy()
y_train = training_data["Capacitance ($\\mu F / cm^2$)"].to_numpy()

X_test = test_data.drop("Capacitance ($\\mu F / cm^2$)", axis=1).to_numpy()
y_test = test_data["Capacitance ($\\mu F / cm^2$)"].to_numpy()

feature_names = X_train_pd.columns.tolist()
training_data
```

```python
def standardise(X, X_train=None):
    """Standardise features.

    Parameters:
        X (np.array): Feature matrix.
        X_train_ (np.array): An optional feature matrix to compute the statistics
            from before applying it to X. If None, just use X to compute the statistics.

    Returns:
        X_std (np.array): Standardised feature matrix
    """
    if X_train is None:
        X_train = X

    mu = np.mean(X_train, axis=0, keepdims=True)
    sigma = np.std(X_train, axis=0, keepdims=True)
    X_std = (X - mu) / sigma
    return X_std

X_test_std = standardise(X_test, X_train)
X_train_std = standardise(X_train)
```

```python
def r2_score(y_true, y_pred):
    """R^2 score to assess regression performance."""

    # Adjustment to avoid subtraction between (K,) and (1, K) arrays.
    y_true = y_true.reshape(y_pred.shape)
    y_bar = y_true.mean()

    ss_tot = ((y_true - y_bar)**2).sum()
    ss_res = ((y_true - y_pred)**2).sum()
    return 1 - (ss_res/ss_tot)

def mse(y_true, y_pred):
    return np.sum((y_true - y_pred) ** 2)

def mse_loss(y_true, y_pred):
    y_true = y_true.reshape(y_pred.shape)
    return np.mean((y_true - y_pred) ** 2) 
```

### (1.1.1) [(index)](#index-task-111)

```python
def split_samples(X, y, col, val):
    left_mask = (X[:, col] < val)

    # Using the binary masks `left_mask`, we split X and y.
    X_l, y_l = X[left_mask, :], y[left_mask]
    X_r, y_r = X[~left_mask, :], y[~left_mask]

    return (X_l, y_l), (X_r, y_r)
```

```python
def split_value(X, y, column):
    """
    Calculate the GINI-index based on `column` with the split that minimizes the GINI-index.
    Arguments:
        X: training features, of shape (N, p).
        y: vector of training labels, of shape (N,).
        sample_weights: weights for each samples, of shape (N,).
        column: the column of the feature for calculating. 0 <= column < D
    Returns:
        (float, float): the resulted GINI-index and the corresponding value used in splitting.
    """

    unique_vals = np.unique(X[:, column])

    index_val, threshold = np.inf, None

    # split the values of i-th feature and calculate the cost
    for value in unique_vals:
        (X_l, y_l), (X_r, y_r) = split_samples(X, y, column, value)

        # if one of the two sides is empty, skip this split.
        if len(y_l) == 0 or len(y_r) == 0:
            continue
        
        new_cost = mse(y_l, np.mean(y_l)) + mse(y_r, np.mean(y_r))
        if new_cost < index_val:
              index_val, threshold = new_cost, value

    return index_val, threshold
```

```python
def split(X, y):
    """
    Choose the best feature to split according to criterion.
    Args:
        X: training features, of shape (N, p).
        y: vector of training labels, of shape (N,).
        sample_weights: weights for each samples, of shape (N,).
        columns_dict: a dictionary mapping column indices to whether the column is categorical or numerical variable.
    Returns:
        (int, float): the best feature index and value used in splitting.
        If the feature index is None, then no valid split for the current Node.
    """

    # Initialize `split_column` to None, so if None returned this means there is no valid split at the current node.
    min_index_val = np.inf
    split_column = None
    split_val = np.nan

    for idx, column in enumerate(X.T):
        # skip column if samples are not seperable by that column.
        if len(np.unique(column)) < 2:
            continue
        index_val, current_val = split_value(X, y, idx)


        if index_val < min_index_val:
            # Keep track with:

            # 1. the current minimum gini-index value,
            min_index_val = index_val

            # 2. corresponding column,
            split_column = idx

            # 3. corresponding split threshold.
            split_val = current_val

    return split_column, split_val
```

```python
def build_tree(X, y, feature_names, depth, max_depth=10, min_samples_leaf=10):

    # include a clause for the cases where (i) no feature, (ii) all labels are the same, (iii) depth exceed, or (iv) X is too small
    if len(np.unique(y))==1 or depth>=max_depth or len(X)<=min_samples_leaf:
        return {'mean_value': np.mean(y)}

    split_index, split_val = split(X, y)

    # If no valid split at this node, use majority vote.
    if split_index is None:
        return {'mean_value': np.mean(y)}

    # Split samples (X, y) given column, split-value, and categorical flag.
    (X_l, y_l), (X_r, y_r) = split_samples(X, y, split_index, split_val)
    return {
        'feature_name': feature_names[split_index],
        'feature_index': split_index,
        'value': split_val,
        'mean_value': None,
        'left': build_tree(X_l, y_l, feature_names, depth + 1, max_depth, min_samples_leaf),
        'right': build_tree(X_r, y_r, feature_names, depth + 1, max_depth, min_samples_leaf)
    }
```

```python
def train(X, y, feature_names):
    return build_tree(X, y, feature_names, depth=1)

trained_dt = train(X_train, y_train, feature_names)
```

```python
def find_predict(tree, x):
    """
    Predict a single sample with the fitted decision tree.
    Args:
        x: ((pd.Dataframe) a single sample features, of shape (D,).
    Returns:
        (int): predicted testing sample label.
    """
    if tree['mean_value'] is not None:
        return tree['mean_value']

    elif x[tree['feature_index']] < tree['value']:
        # go to left branch
        return find_predict(tree['left'], x)
    else:
        # go to right branch
        return find_predict(tree['right'], x)
```

```python
def predict(tree, X):
    """
    Predict classification results for X.
    Args:
        X: (pd.Dataframe) testing sample features, of shape (N, p).
    Returns:
        (np.array): predicted testing sample labels, of shape (N,).
    """
    if len(X.shape) == 1:
        return find_predict(tree, X)
    else:
        return np.array([find_predict(tree, x) for x in X])
```

```python
def tree_score(tree, X_test, y_test):
    y_pred = predict(tree, X_test)
    return (mse_loss(y_test, y_pred),
            r2_score(y_test, y_pred))

dt_mse_train, dt_r2_train = tree_score(trained_dt, X_train, y_train)
dt_mse_test, dt_r2_test = tree_score(trained_dt, X_test, y_test)

print(f"Training: MSE = {dt_mse_train}, R^2 = {dt_r2_train}")
print(f"Test:     MSE = {dt_mse_test}, R^2 = {dt_r2_test}")
```

### (1.1.2) [(index)](#index-task-112)

```python
def split_rf(n_features, X, y):
    """
    Choose the best feature to split according to criterion.
    Args:
        n_features: number of sampled features.
        X: training features, of shape (N, p).
        y: vector of training labels, of shape (N,).
        sample_weights: weights for each samples, of shape (N,).
        columns_dict: a dictionary mapping column indices to whether the column is categorical or numerical variable.
    Returns:
        (float, int, float): the minimized gini-index, the best feature index and value used in splitting.
    """

    # The added sampling step.
    columns = np.random.choice(list(range(X.shape[1])), n_features, replace=False)
    min_index, split_column, split_val = np.inf, 0, np.nan

    # Only scan through the sampled columns in `columns_dict`.
    for column in columns:
        # skip column if samples are not seperable by that column.
        if len(np.unique(X[:, column])) < 2:
            continue

        # search for the best splitting value for the given column.
        index_val, val = split_value(X, y, column)
        if index_val < min_index:
            min_index, split_column, split_val = index_val, column, val

    return min_index, split_column, split_val
```

```python
def build_tree_rf(n_features, X, y, feature_names, depth, max_depth=10, min_samples_leaf=10):
    """Build the decision tree according to the data.
    Args:
        X: (np.array) training features, of shape (N, p).
        y: (np.array) vector of training labels, of shape (N,).
        sample_weights: weights for each samples, of shape (N,).
        columns_dict: a dictionary mapping column indices to whether the column is categorical or numerical variable.
        feature_names (list): record the name of features in X in the original dataset.
        depth (int): current depth for this node.
    Returns:
        (dict): a dict denoting the decision tree (binary-tree). Each node has seven attributes:
          1. 'feature_name': The column name of the split.
          2. 'feature_index': The column index of the split.
          3. 'value': The value used for the split.
          4. 'categorical': indicator for categorical/numerical variables.
          5. 'majority_label': For leaf nodes, this stores the dominant label. Otherwise, it is None.
          6. 'left': The left sub-tree with the same structure.
          7. 'right' The right sub-tree with the same structure.
    """
    # include a clause for the cases where (i) all lables are the same, (ii) depth exceed (iii) X is too small
    if len(np.unique(y)) == 1 or depth>=max_depth or len(X)<=min_samples_leaf:
        return {'mean_value': np.mean(y)}

    else:
        min_index, split_index, split_val = split_rf(n_features, X, y)

        # If GI is infinity, it means that samples are not seperable by the sampled features.
        if min_index == np.inf:
            return {'mean_value': np.mean(y)}

        (X_l, y_l), (X_r, y_r) = split_samples(X, y, split_index, split_val)
        return {
            'feature_name': feature_names[split_index],
            'feature_index': split_index,
            'value': split_val,
            'mean_value': None,
            'left': build_tree_rf(n_features, X_l, y_l, feature_names, depth + 1, max_depth, min_samples_leaf),
            'right': build_tree_rf(n_features, X_r, y_r, feature_names, depth + 1, max_depth, min_samples_leaf)
        }
```

```python
def train_rf(B, n_features, X, y, feature_names, max_depth=10, min_samples_leaf=10):
    """
    Build the decision tree according to the training data.
    Args:
        B: number of decision trees.
        X: (pd.Dataframe) training features, of shape (N, p). Each X[i] is a training sample.
        y: (pd.Series) vector of training labels, of shape (N,). y[i] is the label for X[i], and each y[i] is
        an integer in the range 0 <= y[i] <= C. Here C = 1.
        columns_dict: a dictionary mapping column indices to whether the column is categorical or numerical variable.
        sample_weights: weights for each samples, of shape (N,).
    """
    np.random.seed(0)

    N = X.shape[0]
    training_indices = np.arange(N)
    trees = []
    oobs = []

    for _ in range(B):
        # Sample the training_indices (with replacement)
        sample = np.random.choice(training_indices, N, replace=True)
        oob = list(set(training_indices) - set(sample))
        X_sample = X[sample, :]
        y_sample = y[sample]
        tree = build_tree_rf(n_features, X_sample, y_sample,
                             feature_names, depth=1,
                             max_depth=max_depth, min_samples_leaf=min_samples_leaf)
        trees.append(tree)
        oobs.append(oob)


    return trees, oobs
```

```python
def predict_rf(rf, X):
    """
    Predict classification results for X.
    Args:
        rf: A trained random forest through train_rf function.
        X: (pd.Dataframe) testing sample features, of shape (N, p).
    Returns:
        (np.array): predicted testing sample labels, of shape (N,).
    """

    if len(X.shape) == 1:
        # if we have one sample
        return np.mean([find_predict(tree, X) for tree in rf])
    else:
        # if we have multiple samples
        return np.array([np.mean([find_predict(tree, x) for tree in rf]) for x in X])
```

```python
def rf_score(rf, X_test, y_test):
    y_pred = predict_rf(rf, X_test)
    return (mse_loss(y_test, y_pred),
            r2_score(y_test, y_pred))
```

```python
def cross_validation_score_rf(X, y, B, folds, n_features, max_depth=10, min_samples_leaf=10):
    scores = []
    for fold in folds:
      train_indexes = list(set(range(y_train.shape[0])) - set(fold))

      X_train_i = X_train[train_indexes, :]
      y_train_i = y_train[train_indexes]

      X_val_i = X_train[fold, :]
      y_val_i = y_train[fold]

      trees, _ = train_rf(B, n_features, X_train_i, y_train_i, feature_names,
                          max_depth=max_depth, min_samples_leaf=min_samples_leaf)

      score = mse_loss(predict_rf(trees, X_val_i), y_val_i)
      scores.append(score)
      
    return np.mean(scores)
```

```python
def dense(X, W, b):
    """Full-connected MLP layer.

    Parameters:
        X (np.ndarray): K x h_in array of inputs, where K is the batch size and h_in if the input features dimension.
        W (np.ndarray): h_out x h_in array for kernel matrix parametersm, where h_out is the output dimension.
        b (np.ndarray): Length h_out 1-D array for bias parameters

    Returns:
        a (np.ndarray): K x h_out array of pre-activations
    """
    a = (W @ X.T + b[:, None]).T
    return a
```

```python
# Activation function

def activation(a):
    return np.arctan(a) * np.log(np.abs(a) + 1)

def grad_activation(a):
    grad = (np.log(np.abs(a) + 1) / (1 + a ** 2)
            + np.sign(a) * np.arctan(a) / (np.abs(a) + 1))
    return grad
```

```python
# grad MSE loss
def grad_mse_loss(y_true, y_pred):
    y_true = y_true.reshape(y_pred.shape)
    return 2 * (y_pred - y_true) / y_true.size
```

```python
class MLP:
    def __init__(self, seed=1337):
        self.rng = np.random.default_rng(seed)
        self.layers = []
        self.add_layer(12, 50)
        self.add_layer(50, 50)
        self.add_layer(50, 1)

    def add_layer(self, in_dim, out_dim, activation="identity"):
        """Add fully connected layer to MLP.

        Parameters:
            in_dim (int): The output dimension of the layer.
            out_dim (int): The input dimension of the layer.
            activation (str): The activation function name.
        """
        # store each layer as a dictionary in the list, as shown in the
        # attached diagram.
        self.layers.append({
            # apply Glorot initialisation for weights.
            "W": self.rng.normal(size=(out_dim, in_dim)) * np.sqrt(2. / (in_dim + out_dim)),
            # initialise bias vector with zeros.
            "b": np.zeros(out_dim)
            # store the activation function (as string)
        })

    def predict(self, X):    
        # store pre- and post-activations in list
        forward_pass = [{"index": 0, "a": X, "h": X}]

        # iterate through hidden layers
        for k in range(1, len(self.layers)):
            # compute pre-activations
            a = dense(forward_pass[k - 1]["h"], self.layers[k - 1]["W"], self.layers[k - 1]["b"])
            forward_pass.append({"index": k, "a" : a, "h" : activation(a)})

        y_hat = dense(forward_pass[-1]["h"], self.layers[-1]["W"], self.layers[-1]["b"])
        # predicted target is output of last layer
        return y_hat, forward_pass
    
    def backpropagate(self, forward_pass, delta_output):
        delta = delta_output

        gradients = []
        #Â Create a list that will contain the gradients of all the layers.
        delta = delta_output

        # Iterate on layers backwardly, from output to input.
        # Calculate gradients w.r.t. weights and biases of each level and store in list of dictionaries.
        gradients = []
        for layer, forward_computes in reversed(list(zip(self.layers, forward_pass))):

            h = forward_computes["h"]
            a = forward_computes["a"]

            gradients.append({"W" : delta.T @ h,
                              "b" : delta.sum(axis=0)})

            # Calculate the delta for the backward layer.
            delta = grad_activation(a) * (layer["W"].T @ delta.T).T

        # Return now ordered list matching the layers.
        return list(reversed(gradients))
    
```

```python
def sgd_step(mlp, X, y, learning_rate = 1e-3):

        # Compute the forward pass.
        y_hat, forward_pass = mlp.predict(X)

        # Compute the partial derivative of the loss w.r.t. to predictions `y_hat`.
        delta_output = grad_mse_loss(y, y_hat)

        # Apply backpropagation algorithm to compute the gradients of the MLP parameters.
        gradients = mlp.backpropagate(forward_pass, delta_output)

        for ind, layer in enumerate(mlp.layers):
            mlp.layers[ind]["W"] -= learning_rate * gradients[ind]["W"]
            mlp.layers[ind]["b"] -= learning_rate * gradients[ind]["b"]
```

```python
def sgd(X_train, y_train, X_test, y_test, mlp,
        learning_rate = 1e-3,
        n_epochs = 10,
        minibatchsize = 8,
        seed = 1337):

    # get random number generator
    rng = np.random.default_rng(seed)

    # compute number of iterations per epoch
    n_iterations = len(y_train) // minibatchsize

    # store losses
    losses_train = []
    losses_test = []

    epochs_bar = tqdm(range(n_epochs))
    for i in epochs_bar:

        # shuffle data
        p = rng.permutation(len(y_train))
        X_train_shuffled = X_train[p]
        y_train_shuffled = y_train[p]

        for j in range(n_iterations):
            # get batch
            X_batch = X_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]
            y_batch = y_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]

            # apply sgd step
            sgd_step(mlp, X_batch, y_batch,
                     learning_rate=learning_rate)

        # compute loss at the end of each epoch
        y_hat_train, _ = mlp.predict(X_train)
        losses_train.append(mse_loss(y_train, y_hat_train).squeeze())
        y_hat_test, _ = mlp.predict(X_test)
        losses_test.append(mse_loss(y_test, y_hat_test).squeeze())
        epochs_bar.set_description(f'train_loss: {losses_train[-1]:.2f}, '
                                   f'test_loss: {losses_test[-1]:.2f}, '
                                   f'train_R^2: {r2_score(y_train, y_hat_train):.2f} '
                                   f'test_R^2: {r2_score(y_test, y_hat_test):.2f} ')
    return mlp, losses_train, losses_test
```

```python
def momentum_step(mlp, X, y, vt,
                  beta = 0, learning_rate = 5e-5):
    # Compute the forward pass.
    y_hat, forward_pass = mlp.predict(X)

    # Compute the partial derivative of the loss w.r.t. to predictions `y_hat`.
    delta_output = grad_mse_loss(y, y_hat)

    # Apply backpropagation algorithm to compute the gradients of the MLP parameters.
    g_t = mlp.backpropagate(forward_pass, delta_output)

    for ind in range(len(mlp.layers)):
        vt[ind]["W"] = beta * vt[ind]["W"] + learning_rate * g_t[ind]["W"]
        vt[ind]["b"] = beta * vt[ind]["b"] + learning_rate * g_t[ind]["b"]
        mlp.layers[ind]["W"] -= vt[ind]["W"]
        mlp.layers[ind]["b"] -= vt[ind]["b"]

    return vt
```

```python
def momentum_sgd(X_train, y_train, X_test, y_test, mlp,
                 beta = 0,
                 learning_rate = 1e-3,
                 n_epochs = 10,
                 minibatchsize = 8,
                 seed = 1337):

    # get random number generator
    rng = np.random.default_rng(seed)

    # compute number of iterations per epoch
    n_iterations = len(y_train) // minibatchsize

    # store losses
    losses_train = []
    losses_test = []

    epochs_bar = tqdm(range(n_epochs))
    for i in epochs_bar:

        vt = []
        for layer in mlp.layers:
            vt.append({
                "W" : np.zeros_like(layer["W"]),
                "b" : np.zeros_like(layer["b"])
            })

        # shuffle data
        p = rng.permutation(len(y_train))
        X_train_shuffled = X_train[p]
        y_train_shuffled = y_train[p]

        for j in range(n_iterations):
            # get batch
            X_batch = X_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]
            y_batch = y_train_shuffled[j * minibatchsize
                                       : (j + 1) * minibatchsize]

            # apply sgd step
            vt = momentum_step(mlp, X_batch, y_batch, vt,
                               beta = beta,
                               learning_rate=learning_rate)

        # compute loss at the end of each epoch
        y_hat_train, _ = mlp.predict(X_train)
        losses_train.append(mse_loss(y_train, y_hat_train).squeeze())
        y_hat_test, _ = mlp.predict(X_test)
        losses_test.append(mse_loss(y_test, y_hat_test).squeeze())
        epochs_bar.set_description(f'train_loss: {losses_train[-1]:.2f}, '
                                   f'test_loss: {losses_test[-1]:.2f}, '
                                   f'train_R^2: {r2_score(y_train, y_hat_train):.2f} '
                                   f'test_R^2: {r2_score(y_test, y_hat_test):.2f} ')
    return mlp, losses_train, losses_test
```

```python
def ridge_estimate(X, y, penalty):

    # X: N x D matrix of training inputs
    # y: N x 1 vector of training targets/observations
    # returns: maximum likelihood parameters (D x 1)

    N, D = X.shape
    X_aug = np.hstack([np.ones((N,1)), X]) # augmented training inputs of size N x (D+1)
    N_aug, D_aug = X_aug.shape
    I = np.identity(D_aug)
    I[0, 0] = 0
    beta_ridge = np.linalg.solve(X_aug.T @ X_aug + penalty * I, X_aug.T @ y)
    return beta_ridge
```

```python
def score(X_test, y_test, betas, score_func=mse_loss):
  X_test_aug = np.hstack([np.ones((X_test.shape[0],1)), X_test])
  y_pred = X_test_aug @ betas
  return score_func(y_test, y_pred)
```

```python
def cross_validation_score(X_train, y_train, folds, penalty):
  scores = []
  for i in range(len(folds)):
    val_indexes = folds[i]
    train_indexes = list(set(range(y_train.shape[0])) - set(val_indexes))
    
    X_train_i = X_train[train_indexes, :]
    y_train_i = y_train[train_indexes]

    X_val_i = X_train[val_indexes, :]
    y_val_i = y_train[val_indexes]

    beta_ridge = ridge_estimate(X_train_i, y_train_i, penalty)
    score_i = score(X_val_i, y_val_i,
                    beta_ridge)

    scores.append(score_i)

  # Return the average score
  return np.mean(scores)
```

```python
def choose_best_k(X_train, y_train, folds, penalties):
  penalty_scores = np.zeros(len(penalties))
  
  for i, penalty in enumerate(penalties):
    penalty_scores[i] = cross_validation_score(X_train, y_train, folds, penalty)

  best_pen_idx = np.argmin(penalty_scores)
  return (penalties[best_pen_idx], penalty_scores.min()), penalty_scores
```

```python
def euc_dist(u, v):
    return np.sqrt(np.sum((u - v) ** 2, axis=1))

def hamming(u, v):
    return (u != v).astype(int)

def combined_dist(X_train, x_test):
    euc = euc_dist(np.delete(X_train, cat_col, 1),
                   np.delete(x_test, cat_col))
    ham = hamming(X_train[:, cat_col], x_test[cat_col])
    return euc + ham
```

```python
def k_neighbours(X_train, X_test, k=5):
    n_neighbours = k
    neigh_ind = []

    # compute distance from each point x_text in X_test to all points in X_train
    point_dist = [combined_dist(X_train, x_test) for x_test in X_test]
    
    neigh_ind = np.argsort(point_dist, axis=1)[:,:k]

    return np.array(neigh_ind)
```

```python
def predict(X_train, y_train, X_test, k=5):
  neighbours = k_neighbours(X_train, X_test, k=k)
  y_pred =  np.array([np.argmax(np.bincount(y_train[neighbour]))
                      for neighbour in neighbours])
  return y_pred
```

```python
def class_qualities(y_true, y_pred):
  classes = [0, 1, 2]
  accuracies = np.zeros(3)
  precisions = np.zeros(3)
  N = y_true.size

  for c in classes:
    true_c = y_true[y_true == c].size
    pred_c = y_pred[y_pred == c].size
    matches = y_true[y_pred == y_true]
    TP = matches[matches == c].size
    FP = pred_c - TP
    FN = true_c - TP
    TN = y_true.size - TP - FP - FN
    accuracies[c] = (TP + TN) / N # TP / true_c
    if TP == 0 and FP == 0:
      precisions[c] = np.nan
    else:
      precisions[c] = TP / (TP + FP)

  return accuracies, precisions

def accuracy(y_true, y_pred):
  class_acc, _ = class_qualities(y_true, y_pred)
  N = y_true.size
  micro = np.sum(y_true == y_pred) / N
  weights = np.bincount(y_test) / N
  weighted = np.sum(class_acc * weights)

  # Averaging all accuracies for each class we obtain,
  macro = 1/3 + 2/3 * micro
  return micro, macro, weighted

def precision(y_true, y_pred):
  _, class_prc = class_qualities(y_true, y_pred)
  N = y_true.size
  micro = np.sum(y_true == y_pred) / N
  weights = np.bincount(y_true) / N
  macro = np.mean(class_prc)
  weighted = np.sum(class_prc * weights)
  return micro, macro, weighted

def score(X_train, y_train, X_test, y_test, k=5):
  y_pred = predict(X_train, y_train, X_test, k)
  score, _, _ = accuracy(y_test, y_pred)
  return score
```

```python
def cross_validation_score(X_train, y_train, folds, k):
    scores = []
    for i in range(len(folds)):
        val_indexes = folds[i]
        train_indexes = list(set(range(y_train.shape[0])) - set(val_indexes))

        X_train_i = X_train[train_indexes, :]
        y_train_i = y_train[train_indexes]

        X_val_i = X_train[val_indexes, :]
        y_val_i = y_train[val_indexes]

        X_val_i = standardise(X_val_i, X_train=X_train_i)
        X_train_i = standardise(X_train_i)

        score_i = score(X_train_i, y_train_i, X_val_i, y_val_i, k=k)

        scores.append(score_i)

    return np.mean(scores)
```

```python
def choose_best_k(X_train, y_train, folds, k_range):
    k_scores = np.zeros((len(k_range),))
  
    for i, k in enumerate(k_range):
        k_scores[i] = cross_validation_score(X_train, y_train, folds, k)

    best_k_index = np.argmax(k_scores)
    return k_range[best_k_index], k_scores
```

```python
def two_step_predict(X_train, y_train, X_test, k=5):
    malignant = (y_train != 0)
    neighbours = k_neighbours(X_train, X_test, k=k)
    y_combined = y_train.copy()
    y_combined[y_train != 0] = 1

    y_pred =  np.array([np.argmax(np.bincount(y_combined[neighbour], minlength=3))
                        for neighbour in neighbours])

    non_zero_pred = (y_pred != 0)

    if non_zero_pred.any():
        neighbours = k_neighbours(X_train[malignant], X_test[non_zero_pred], k=1)

        y_pred_mal = np.array([np.argmax(np.bincount(y_train[malignant][neighbour]))
                                    for neighbour in neighbours])

        y_pred[non_zero_pred] = y_pred_mal
    return y_pred
```

```python
def two_step_qualities(X_train, y_train, X_test, y_test, k=5):
    y_pred = two_step_predict(X_train, y_train, X_test, k)
    micro_acc, macro_acc, _ = accuracy(y_test, y_pred)
    micro_prc, macro_prc, _ = precision(y_test, y_pred)
    return (micro_acc, macro_acc), (micro_prc, macro_prc)
```

## (2.2) Logistic regression vs kernel logistic regression [(index)](#index-task-22)

### (2.2.1) [(index)](#index-task-221)

```python
def loss(X, y, beta, beta0, lam=0.0025):
    y_log = h(X, beta, beta0)
    return (-np.mean(y * np.log(y_log)
                     + (1 - y) * np.log(1 - y_log))
            + lam / 2 * np.linalg.norm(beta) ** 2)

def h(X, beta, beta0):
    return 1 / (1 + np.exp(-(X @ beta + beta0)))

def grad_loss(X, y, beta, beta0, lam=0.0025):
    y_log = h(X, beta, beta0)
    beta_loss = np.mean(X.T * (y_log - y), axis=1) + lam * beta
    beta0_loss = np.mean(y_log - y)
    return beta_loss, beta0_loss
```

```python
def propagate(X, y, beta, beta0):
    # Mean sample loss function
    mean_loss = loss(X, y, beta, beta0)

    # Derivatives
    beta_loss, beta0_loss = grad_loss(X, y, beta, beta0)
    mean_loss = np.squeeze(mean_loss)

    return beta_loss, beta0_loss, mean_loss
```

```python
def optimise(X, y, beta, beta0, num_iterations=1000, learning_rate=0.1, print_loss=False):
    mean_loss_history = []
    
    for i in range(num_iterations):

        # Calculating the loss and gradients
        dbeta, dbeta0, mean_loss = propagate(X, y, beta, beta0)
      
        # Updating procedure
        beta -= learning_rate * dbeta
        beta0 -= learning_rate * dbeta0
      
        # Record the loss values
        if i % 10 == 0:
            mean_loss_history.append(mean_loss)
      
        # Printing the loss every 100 iterations
        if print_loss and i % 100 == 0:
            print ('Mean loss after iteration %i: %f' %(i, mean_loss))

    # Saving parameters and gradients in dictionary
    params = {'betas': (beta, beta0)}
    grads = {'dbeta': (dbeta, dbeta0)}
  
    return params, grads, mean_loss_history
```

```python
def model(X_train, y_train, X_test, y_test, num_iterations=2000, learning_rate=0.1, print_loss=False):
    # Initialising parameters with zeros
    beta, beta0 = np.zeros(X_train.shape[1]), 0.
    # last element is beta0

    # Gradient descent
    parameters, grads, mean_loss_history = optimise(X_train, y_train, beta, beta0,
                                                    num_iterations, learning_rate, print_loss=print_loss)

    # Retrieving parameter vector beta from dictionary 'parameters'
    beta, beta0 = parameters['betas']

    # Predicting test and training set examples
    y_pred_test = np.round(h(X_test, beta, beta0))
    y_pred_train = np.round(h(X_train, beta, beta0))

    # Printing train/test accuracy
    print('Training accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_train - y_train)) * 100))
    print('Test accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_test - y_test)) * 100))

    # Saving all the information
    d = {'mean_loss_history': mean_loss_history, 'betas': (beta, beta0),
         'y_pred_test': y_pred_test, 'y_pred_train': y_pred_train}
  
    return d
```

```python
y_train_binary = y_train.copy()
y_test_binary = y_test.copy()
y_train_binary[y_train != 0] = 1
y_test_binary[y_test != 0] = 1

d = model(X_train_std, y_train_binary, X_test_std, y_test_binary, num_iterations=1000, learning_rate=0.1, print_loss=True)
```

```python
def pr_curve(y_true, y_log):
    """
    Arguments:
    y_true: Ground truth labels with size N
    y_log: Probabilities produced by logistic regression model with size N
  
    Returns:
    auc: Area Under the Curve (AUC)
    tpr_l: List of true positive rate (TPR) values for each scanned threshold
    fpr_l: List of false positive rate (FPR) values for each scanned threshold
    thresholds: Scanned thresholds sorted in decreasing order
    
    """
    # List of distinct values in y_log, sorted sorted in decreasing order
    thresholds = reversed(sorted(set(y_log)))
    tpr_l, prc_l = [], []

    for threshold in thresholds:
        # Thresholding
        y_thresholded = (y_log >= threshold)

        # True positives
        tp = np.sum(y_true & y_thresholded)
        # False positives
        fp = np.sum((~y_true) & y_thresholded)
        # False negatives
        fn = np.sum(y_true & (~y_thresholded))

        tpr = tp / (tp + fn)
        prc = tp / (tp + fp)
        tpr_l.append(tpr)
        prc_l.append(prc)

    # Compute AUC using Trapezoidal rule
    auc = np.trapz(prc_l, tpr_l)
    return auc, tpr_l, prc_l, thresholds
```

### (2.2.2) [(index)](#index-task-222)

We set up the problem with the data matrix $\bm{X}\in\mathbb{R}^{N\times p}$ and classification vectors $\bm{y}\in\mathbb{R}^p$. We represent the kernel as $k(\bm{x},\bm{y})=\phi(\bm{x})\cdot\phi(\bm{y})$. We now optimise the kernelised logistic regression in terms of $\bm{\beta}_{\phi(\bm{X})}$, a higher dimension set of parameters on our new set of features, $\phi(\bm{X})$. This allows us to rewrite the loss function as,
\begin{equation*}
    E(L) = -\frac{1}{N}\sum_{i=1}^{N}y^{(i)}\log h_{\bm{\beta}_{\phi(\bm{X})},\beta_0}(\phi(\bm{X}^{(i)}))+(1-y^{(i)})\log(1-h_{\bm{\beta}_{\phi(\bm{X})}, \beta_0}(\phi(\bm{X}^{(i)}))) + \frac{\lambda}{2}\Vert\bm{\beta}_{\phi(\bm{X})}\Vert^2
\end{equation*}
With the function $h_{\bm{\beta},\beta_0}(\bm{x})=h(\bm{\beta},\beta_0,\bm{x})$ as defined as before (Here, $\bm{X}^{(i)}$ is row $i$ of matrix $\bm{X}$). Now if we introduce a vector $\bm{u}\in\mathbb{R}^N$ such that,
\begin{equation*}
    \bm{\beta}_{\phi(\bm{X})} = \sum_{i=1}^N\phi(\bm{X}^{(i)})\bm{u}_i
\end{equation*}
We have the simplifications,
\begin{equation*}
    \phi(\bm{X}^{(i)})^T\bm{\beta}_{\phi(\bm{X})}=\bm{K}^{(i)}\bm{u},\qquad \Vert\bm{\beta}_{\phi(\bm{X})}\Vert^2=\bm{\beta}_{\phi(\bm{X})}\cdot\bm{\beta}_{\phi(\bm{X})}=\bm{u}^T\bm{Ku}
\end{equation*}
where $\bm{K}$ is the Gram matrix with entries, $\bm{K}_{ij}=\phi(\bm{X}^{(i)})\cdot\phi(\bm{X}^{(j)})=k(\bm{X}^{(i)},\bm{X}^{(j)})$. This lets us write the optimisation problem in terms of $\bm{K}$ while optimising $\bm{u}$ and $\beta_0$,
\begin{equation*}
    E(L) = -\frac{1}{N}\sum_{i=1}^{N}y^{(i)}\log h_{\bm{u},\beta_0}(\bm{K}^{(i)T})+(1-y^{(i)})\log(1-h_{\bm{u}, \beta_0}(\bm{K}^{(i)T})) + \frac{\lambda}{2}\bm{u}^T\bm{Ku}
\end{equation*}
We can differentiate this with respect to $\bm{u}$ and $\beta_0$ in a similar fashion to the non-kernelised logistic regression to obtain derivatives,
\begin{align*}
    \nabla_{\bm{u}}E(L) &= \frac{1}{N}\sum_{i=1}^N\left((h_{\bm{u}, \beta_0}(\bm{K}^{(i)T})-y^{(i)})\bm{K}^{(i)T}\right) + \lambda\bm{Ku} \\
    \nabla_{\beta_0}E(L) &= \frac{1}{N}
\end{align*}

```python
def kernel_matrix(X1, X2, a):
    n, _ = X1.shape
    m, _ = X2.shape
    kernel = np.zeros((n, m))
    
    for ind, x1 in enumerate(X1):
        val = np.exp(-a * np.linalg.norm(X2 - x1, ord=1, axis=1))
        kernel[ind, :] = val
    return kernel

def kernel_loss(u, K, y, beta0, lam=0.0025):
    k_y_log = h(K, u, beta0)
    return (-np.mean(y * np.log(k_y_log)
                     + (1 - y) * np.log(1 - k_y_log))
            + lam / 2 * u.T @ K @ u)

def grad_kernel_loss(u, K, y, beta0, lam=0.0025):
    # grad w.r.t u, beta0

    k_y_log = h(K, u, beta0)
    beta_loss = np.mean(K.T * (k_y_log - y), axis=1) + lam * K @ u
    beta0_loss = np.mean(k_y_log - y)
    return beta_loss, beta0_loss
```

### (2.2.3) [(index)](#index-task-223)

```python
def kernel_optimise(u, K, y, beta0, num_iterations=1000, learning_rate=0.1, print_loss=False):
    mean_loss_history = []
    
    for i in range(num_iterations):

        # Calculating the loss and gradients
        du, dbeta0 = grad_kernel_loss(u, K, y, beta0)
        mean_loss = np.squeeze(kernel_loss(u, K, y, beta0))
      
        # Updating procedure
        u -= learning_rate * du
        beta0 -= learning_rate * dbeta0
      
        # Record the loss values
        if i % 10 == 0:
            mean_loss_history.append(mean_loss)
      
        # Printing the loss every 100 iterations
        if print_loss and i % 100 == 0:
            print ('Mean loss after iteration %i: %f' %(i, mean_loss))

    # Saving parameters and gradients in dictionary
    params = (u, beta0)
    grads = (du, dbeta0)
  
    return params, grads, mean_loss_history
```

```python
def kernel_model(X_train, y_train, X_test, y_test,
                 alpha=0.3, num_iterations=2000,
                 learning_rate=0.1, print_loss=False):
    # Initialising parameters with zeros
    K = kernel_matrix(X_train, X_train, alpha)
    u, beta0 = np.zeros(K.shape[1]), 0.

    # Gradient descent
    parameters, grads, mean_loss_history = kernel_optimise(u, K, y_train, beta0,
                                                           num_iterations, learning_rate,
                                                           print_loss=print_loss)

    # Retrieving parameter vector beta from dictionary 'parameters'
    u, beta0 = parameters

    # Predicting test and training set examples
    K_test = kernel_matrix(X_test, X_train, alpha)
    y_pred_test = np.round(h(K_test, u, beta0))
    y_pred_train = np.round(h(K, u, beta0))

    # Printing train/test accuracy
    print('Training accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_train - y_train)) * 100))
    print('Test accuracy: {} %'.format(100 - np.mean(np.abs(y_pred_test - y_test)) * 100))

    # Saving all the information
    d = {'mean_loss_history': mean_loss_history, 'params': (u, beta0),
         'y_pred_test': y_pred_test, 'y_pred_train': y_pred_train}
  
    return d
```
